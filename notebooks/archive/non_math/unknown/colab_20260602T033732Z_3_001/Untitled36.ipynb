```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
x0 = 1.0
c_j = 0.j
c_k = 1.0 +j
d = c_k - c_j

# tau en R_+
tau = np.linspace(0.001, 8.0, 3000)

# Kernel cusp-cusp
K = (1/(np.pi*1j)) * 1/(tau - x0 + 1j*d)

plt.figure(figsize=(6, 6))

# Curva del kernel en el plano complejo
plt.plot(K.real, K.imag, linewidth=2)

# Punto inicial y punto cercano a tau = x
idx_x = np.argmin(np.abs(tau - x0))
plt.scatter(K.real[0], K.imag[0], label=r'$\tau \approx 0$')
plt.scatter(K.real[idx_x], K.imag[idx_x], label=r'$\tau=x$')
plt.scatter(K.real[-1], K.imag[-1], label=r'$\tau$ grande')

# Ejes real e imaginario
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)

plt.xlabel(r'$\operatorname{Re} K_x(\tau)$')
plt.ylabel(r'$\operatorname{Im} K_x(\tau)$')
plt.title(fr'Kernel cusp-cusp en el plano complejo: $x={x0}$, $c_j={c_j}$, $c_k={c_k}$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# -----------------------------
# Parámetros del modelo cusp-cusp
# -----------------------------
c_j = 0.0
c_k = 1.0
d = c_k - c_j

# Variable tau en R_+
tau = np.linspace(0.001, 8.0, 3000)

# Valores de x que van a variar
x_values = np.linspace(0.05, 5.0, 120)

# -----------------------------
# Función kernel
# -----------------------------
def kernel_cusp_cusp(tau, x, d):
    return (1 / (np.pi * 1j)) * 1 / (tau - x + 1j * d)

    # -----------------------------
    # Circunferencia teórica completa
    # -----------------------------
    if d == 0:
        raise ValueError("d = c_k - c_j no puede ser cero para esta visualización circular.")

        center = -1 / (2 * np.pi * d)
        radius = abs(1 / (2 * np.pi * d))

        theta = np.linspace(0, 2*np.pi, 1000)
        circle_x = center + radius * np.cos(theta)
        circle_y = radius * np.sin(theta)

        # -----------------------------
        # Configuración de figura
        # -----------------------------
        fig, ax = plt.subplots(figsize=(6, 6))

        # Circunferencia completa de referencia
        ax.plot(circle_x, circle_y, linestyle="--", linewidth=1.2, label="Circunferencia teórica")

        # Curva dinámica del kernel
        line, = ax.plot([], [], linewidth=2.5, label=r"$K_x(\tau)$, $\tau\in\mathbb R_+$")

        # Puntos especiales
        p_tau0 = ax.scatter([], [], s=45, label=r"$\tau\approx 0$")
        p_taux = ax.scatter([], [], s=45, label=r"$\tau=x$")
        p_tauinf = ax.scatter([], [], s=45, label=r"$\tau$ grande")

        # Texto dinámico
        text = ax.text(
            0.03, 0.95, "",
                transform=ax.transAxes,
                    fontsize=11,
                        verticalalignment="top"
                        )

                        # Ejes real e imaginario
                        ax.axhline(0, linewidth=1)
                        ax.axvline(0, linewidth=1)

                        ax.set_xlabel(r"$\operatorname{Re} K_x(\tau)$")
                        ax.set_ylabel(r"$\operatorname{Im} K_x(\tau)$")
                        ax.set_title("Kernel cusp-cusp en el plano complejo variando x")
                        ax.grid(True, alpha=0.3)
                        ax.legend(loc="lower right")
                        ax.set_aspect("equal", adjustable="box")

                        # Límites fijos para que no brinque la animación
                        margin = 0.08
                        ax.set_xlim(center - radius - margin, center + radius + margin)
                        ax.set_ylim(-radius - margin, radius + margin)

                        # -----------------------------
                        # Inicialización
                        # -----------------------------
                        def init():
                            line.set_data([], [])
                                p_tau0.set_offsets(np.empty((0, 2)))
                                    p_taux.set_offsets(np.empty((0, 2)))
                                        p_tauinf.set_offsets(np.empty((0, 2)))
                                            text.set_text("")
                                                return line, p_tau0, p_taux, p_tauinf, text

                                                # -----------------------------
                                                # Animación
                                                # -----------------------------
                                                def update(frame):
                                                    x0 = x_values[frame]
                                                        K = kernel_cusp_cusp(tau, x0, d)

                                                            line.set_data(K.real, K.imag)

                                                                # Punto tau cercano a 0
                                                                    p_tau0.set_offsets([[K.real[0], K.imag[0]]])

                                                                        # Punto tau cercano a x
                                                                            idx_x = np.argmin(np.abs(tau - x0))
                                                                                p_taux.set_offsets([[K.real[idx_x], K.imag[idx_x]]])

                                                                                    # Punto tau grande
                                                                                        p_tauinf.set_offsets([[K.real[-1], K.imag[-1]]])

                                                                                            text.set_text(
                                                                                                    fr"$x={x0:.2f}$" "\n"
                                                                                                            fr"$c_j={c_j}$, $c_k={c_k}$" "\n"
                                                                                                                    fr"$d=c_k-c_j={d}$"
                                                                                                                        )

                                                                                                                            return line, p_tau0, p_taux, p_tauinf, text

                                                                                                                            ani = FuncAnimation(
                                                                                                                                fig,
                                                                                                                                    update,
                                                                                                                                        frames=len(x_values),
                                                                                                                                            init_func=init,
                                                                                                                                                blit=True,
                                                                                                                                                    interval=80
                                                                                                                                                    )

                                                                                                                                                    # Guardar como GIF
                                                                                                                                                    ani.save("kernel_cusp_cusp_variando_x.gif", writer=PillowWriter(fps=15))

                                                                                                                                                    plt.show()
```
