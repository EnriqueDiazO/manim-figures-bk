```python
from turtle import *
import colorsys
import math

speed(0.25)
bgcolor("black")
goto(0, -40)
h = 0

for i in range(16):
    for j in range(18):
        c = colorsys.hsv_to_rgb(0.125, j / 18, 1)  # Tonos amarillos
        color(c)
        rt(90)
        circle(150 - j * 6, 90)
        lt(90)
        circle(150 - j * 6, 90)
        rt(180)
        circle(49, 24)

done()
```
