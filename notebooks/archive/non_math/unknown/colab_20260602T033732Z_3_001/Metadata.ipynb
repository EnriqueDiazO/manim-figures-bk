```python
import os
import datetime

file_path = "/content/sample_data/california_housing_test.csv"

# Get file status information
stat_info = os.stat(file_path)

# Extract common metadata
file_size = stat_info.st_size  # Size in bytes
creation_time = datetime.datetime.fromtimestamp(stat_info.st_ctime)
modification_time = datetime.datetime.fromtimestamp(stat_info.st_mtime)
access_time = datetime.datetime.fromtimestamp(stat_info.st_atime)

print(f"File Size: {file_size} bytes")
print(f"Creation Time: {creation_time}")
print(f"Modification Time: {modification_time}")
print(f"Access Time: {access_time}")
```
