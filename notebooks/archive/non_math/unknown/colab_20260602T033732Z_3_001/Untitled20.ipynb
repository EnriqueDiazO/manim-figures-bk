```python
import numpy as np
import matplotlib.pyplot as plt

# Tamaño de la imagen (puedes cambiar a 2000x1200 si quieres más resolución)
width, height = 400, 240

# Crear mallas de coordenadas normalizadas
m = np.linspace(0, width - 1, width)
n = np.linspace(0, height - 1, height)
M, N = np.meshgrid(m, n)

# Coordenadas centradas y normalizadas
x = (M - width / 2) / (width / 2)
y = (N - height / 2) / (height / 2)

# Variables polares
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

# Función que simula un vórtice con ondas
z = np.sin(10 * r - 5 * theta)

# Normalizar valores al rango [0, 1]
z_norm = (z - z.min()) / (z.max() - z.min())

# Generar imagen en escala de grises azulada
img = plt.cm.Blues(z_norm)

# Mostrar la imagen
plt.figure(figsize=(10, 6))
plt.imshow(img, origin='lower')
plt.axis('off')
plt.title("Simulación estilo 'Eye of a Storm'")
plt.tight_layout()
plt.show()
```
