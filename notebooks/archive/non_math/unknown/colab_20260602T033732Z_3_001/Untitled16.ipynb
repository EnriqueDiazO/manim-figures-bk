```python
from sympy import sin, cos, pi, symbols
from sympy.plotting import plot_parametric

# Variable simbólica
t = symbols('t')

# Primera curva (32)
r1 = sin(t*32)
p1 = plot_parametric(
    r1*sin(t), r1*cos(t), (t, 0, 2*pi),
    line_color='red',
    show=False,
    label=r"$r = \sin(32\theta)$"
)

# Segunda curva (44)
r2 = sin(t*44)
p2 = plot_parametric(
    r2*sin(t), r2*cos(t), (t, 0, 2*pi),
    line_color='blue',
    show=False,
    label=r"$r = \sin(44\theta)$"
)

# Combinar en un solo gráfico
p1.extend(p2)
p1.title = "Curvas $r = \sin(32\\theta)$ y $r = \sin(44\\theta)$ en coordenadas cartesianas"
p1.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de t
t = np.linspace(0, 2*np.pi, 6550)

# Primera curva (32)
r1 = np.sin(t*32)
x1 = r1 * np.sin(t)
y1 = r1 * np.cos(t)

# Segunda curva (44)
r2 = np.sin(t*44)
x2 = r2 * np.sin(t)
y2 = r2 * np.cos(t)

# Crear figura
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')

# Colores variables según t
# Primera curva
sc1 = ax.scatter(x1, y1, c=t, cmap='Purples', s=1, label=r"$r = \sin(32\theta)$")
# Segunda curva
sc2 = ax.scatter(x2, y2, c=t, cmap='viridis', s=1, label=r"$r = \sin(44\theta)$")

# Barra de color
cbar = plt.colorbar(sc1, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label("Valor de $t$")

# Título y leyenda
ax.set_title("Curvas $r = \sin(32\\theta)$ y $r = \sin(44\\theta)$ con color según $t$")
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib.colors import Normalize

# === Datos ===
t = np.linspace(0, 2*np.pi, 20000)

# Curva 1: r = sin(32 t)
r1 = np.sin(32*t)
x1 = r1 * np.sin(t)
y1 = r1 * np.cos(t)

# Curva 2: r = sin(44 t)
r2 = np.sin(44*t)
x2 = r2 * np.sin(t)
y2 = r2 * np.cos(t)

# === Config colormaps y normalización compartida de t ===
cmap1 =  get_cmap('RdYlBu')
cmap2 = get_cmap('PiYG')
norm = Normalize(vmin=t.min(), vmax=t.max())

# === Figura ===
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_aspect('equal', adjustable='box')

# Dibujar cada curva con colores dependientes de t
sc1 = ax.scatter(x1, y1, c=t, cmap=cmap1, norm=norm, s=2, linewidths=0, label=r"$r=\sin(32\theta)$")
sc2 = ax.scatter(x2, y2, c=t, cmap=cmap2, norm=norm, s=2, linewidths=0, label=r"$r=\sin(44\theta)$")

# Título y ejes
ax.set_title(r"Curvas $r=\sin(32\theta)$ y $r=\sin(44\theta)$ coloreadas por $t$")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Leyenda simple para identificar cada curva
ax.legend(loc="upper right", frameon=True)

# Dos barras de color (una por colormap), mismo rango de t
cbar1 = fig.colorbar(sc1, ax=ax, fraction=0.046, pad=0.04)
cbar1.set_label(r"$t$ (para $r=\sin(32\theta)$)")

cbar2 = fig.colorbar(sc2, ax=ax, fraction=0.046, pad=0.10)
cbar2.set_label(r"$t$ (para $r=\sin(44\theta)$)")

# Opcional: estética
ax.grid(False)

plt.tight_layout()
plt.show()
```
