```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios de una distribución normal
data = np.random.normal(loc=2, scale=2, size=1000000)

# Calcular los cuantiles
quantiles = np.linspace(0, 1, 10000)
quantile_values = np.quantile(data, quantiles)

# Graficar la función cuantil
plt.plot(quantiles, quantile_values, label="Función cuantil")
plt.xlabel("Cuantil")
plt.ylabel("Valor")
plt.title("Función cuantil de una distribución normal")
plt.grid(True)
plt.legend()
plt.show()
```
