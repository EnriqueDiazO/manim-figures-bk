```python
from pathlib import Path
import os, stat, datetime, pwd, grp

def get_file_metadata(filepath: str) -> dict:
    p = Path(filepath)
    if not p.exists():
        raise FileNotFoundError(filepath)
    s = p.stat()
    return {
        "path": str(p.resolve()),
        "name": p.name,
        "suffix": p.suffix,
        "size_bytes": s.st_size,
        "modified": datetime.datetime.fromtimestamp(s.st_mtime).isoformat(),
        "created": datetime.datetime.fromtimestamp(s.st_ctime).isoformat(),
        "permissions": stat.filemode(s.st_mode),
        "owner": pwd.getpwuid(s.st_uid).pw_name if hasattr(pwd, 'getpwuid') else s.st_uid,
        "group": grp.getgrgid(s.st_gid).gr_name if hasattr(grp, 'getgrgid') else s.st_gid,
        "is_symlink": p.is_symlink(),
        "is_dir": p.is_dir(),
        "is_file": p.is_file()
    }

    get_file_metadata("/content/sample_data/california_housing_test.csv")
```
