```python
import numpy as np
import scipy.stats as stats

# Definir los datos
f1 = np.array([7, 29, 31, 49, 39, 29, 11, 4, 1, 0])  # Añadí el 0 al final para igualar el tamaño de f2
f2 = np.array([4, 17, 45, 78, 128, 96, 75, 39, 16, 2])

# Crear la tabla combinada
TBL = np.vstack([f1, f2])

# Realizar la prueba chi-cuadrado
chi2_stat, p_val, dof, expected = stats.chi2_contingency(TBL)

# Mostrar resultados
print(f"Chi-cuadrado: {chi2_stat}")
print(f"Grados de libertad: {dof}")
print(f"P-valor: {p_val}")

# Advertencia sobre celdas pequeñas
if np.any(TBL < 5):
    print("Advertencia: La aproximación de chi-cuadrado puede ser incorrecta debido a celdas pequeñas.")

# Simulación del valor P con 2000 replicados
simulated_p_vals = []
for _ in range(2000):
    simulated_f1 = np.random.multinomial(sum(f1), f1/sum(f1))
    simulated_f2 = np.random.multinomial(sum(f2), f2/sum(f2))
    simulated_TBL = np.vstack([simulated_f1, simulated_f2])
    chi2_stat_sim, _, _, _ = stats.chi2_contingency(simulated_TBL)
    simulated_p_vals.append(chi2_stat_sim)

simulated_p_val = np.mean(np.array(simulated_p_vals) >= chi2_stat)
print(f"P-valor simulado (2000 réplicas): {simulated_p_val}")
```

```python
import numpy as np
import scipy.stats as stats

# Crear cuatro series de datos continuos como ejemplo
x1 = np.random.normal(loc=0, scale=1, size=500)  # Serie de datos 1
x2 = np.random.normal(loc=1, scale=1.5, size=500)  # Serie de datos 2
x3 = np.random.normal(loc=0.5, scale=1.2, size=500)  # Serie de datos 3
x4 = np.random.normal(loc=1.2, scale=1.1, size=500)  # Serie de datos 4

# Crear los histogramas (sin graficarlos) para obtener las cuentas en cada intervalo (bin)
f1_counts, bin_edges_1 = np.histogram(x1, bins=5)  # Histograma para x1
f2_counts, _ = np.histogram(x2, bins=bin_edges_1)   # Histograma para x2 usando los mismos bins

f3_counts, bin_edges_2 = np.histogram(x3, bins=5)  # Histograma para x3
f4_counts, _ = np.histogram(x4, bins=bin_edges_2)   # Histograma para x4 usando los mismos bins

# Crear las dos tablas de contingencia
TBL1 = np.vstack([f1_counts, f2_counts])  # Tabla de contingencia para x1 vs x2
TBL2 = np.vstack([f3_counts, f4_counts])  # Tabla de contingencia para x3 vs x4

# Sumar las tablas de contingencia para formar una tabla combinada
TBL_combined = TBL1 + TBL2

# Realizar la prueba chi-cuadrado para la tabla combinada
chi2_stat_combined, p_val_combined, dof_combined, expected_combined = stats.chi2_contingency(TBL_combined)

# Mostrar resultados
print(f"Chi-cuadrado combinado: {chi2_stat_combined}")
print(f"Grados de libertad combinados: {dof_combined}")
print(f"P-valor combinado: {p_val_combined}")

# Simulación del valor P con 2000 replicados
simulated_p_vals_combined = []
for _ in range(2000):
    simulated_f1_combined = np.random.multinomial(sum(TBL_combined[0]), TBL_combined[0]/sum(TBL_combined[0]))
    simulated_f2_combined = np.random.multinomial(sum(TBL_combined[1]), TBL_combined[1]/sum(TBL_combined[1]))
    simulated_TBL_combined = np.vstack([simulated_f1_combined, simulated_f2_combined])
    chi2_stat_sim_combined, _, _, _ = stats.chi2_contingency(simulated_TBL_combined)
    simulated_p_vals_combined.append(chi2_stat_sim_combined)

simulated_p_val_combined = np.mean(np.array(simulated_p_vals_combined) >= chi2_stat_combined)
print(f"P-valor simulado combinado (2000 réplicas): {simulated_p_val_combined}")
```
