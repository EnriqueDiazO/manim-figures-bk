```python
import nltk
import pyphen

# Descargar el corpus de CESS-ESP
nltk.download('cess_esp')
palabras = nltk.corpus.cess_esp.words()

# Inicializar el diccionario de sílabas para español
diccionario = pyphen.Pyphen(lang='es')

def contar_silabas(palabra):
    silabas = diccionario.inserted(palabra.lower()).split('-')
    return len(silabas)

# Filtrar palabras alfabéticas y contar sílabas
silabas_por_palabra = [contar_silabas(palabra) for palabra in palabras if palabra.isalpha()]

# Calcular el promedio
promedio_silabas = sum(silabas_por_palabra) / len(silabas_por_palabra)
print(f"Promedio de sílabas por palabra: {promedio_silabas:.2f}")
```
