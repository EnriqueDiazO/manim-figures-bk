```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Circle, Ellipse
import matplotlib as mpl

# Figure
fig, ax = plt.subplots(figsize=(12, 8), dpi=180)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# Background
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

# Title box
title = FancyBboxPatch((0.05, 0.9), 0.9, 0.07,
                       boxstyle="round,pad=0.012,rounding_size=0.02",
                       linewidth=1.5, edgecolor="black", facecolor="#dfefff")
ax.add_patch(title)
ax.text(0.5, 0.935, "Separation of a Compact Set and a Closed Set",
        ha="center", va="center", fontsize=18, fontweight="bold")

# Theorem statement
stmt = FancyBboxPatch((0.05, 0.78), 0.9, 0.09,
                      boxstyle="round,pad=0.012,rounding_size=0.02",
                      linewidth=1.0, edgecolor="black", facecolor="#f8f8f8")
ax.add_patch(stmt)
ax.text(0.07, 0.835,
        r"If $K,C\subset X$, $K$ compact, $C$ closed, and $K\cap C=\varnothing$, "
        "\nthen there exists a neighborhood $V$ of $0$ such that"
        "\n" + r"$(K+V)\cap(C+V)=\varnothing$.",
        ha="left", va="center", fontsize=13)

# Rhetorical box
box1 = FancyBboxPatch((0.05, 0.44), 0.28, 0.28,
                      boxstyle="round,pad=0.015,rounding_size=0.02",
                      linewidth=1.2, edgecolor="black", facecolor="#fff4d6")
ax.add_patch(box1)
ax.text(0.07, 0.69, "Rhetorical level", fontsize=15, fontweight="bold")
ax.text(0.07, 0.66,
        "A compact set and a closed set,\n"
        "if disjoint, can be separated\n"
        "uniformly by the same small\n"
        "neighborhood of the origin.\n\n"
        "After “thickening” both sets\n"
        "by that neighborhood, they still\n"
        "do not touch.",
        ha="left", va="top", fontsize=11)

# Symbolic box
box2 = FancyBboxPatch((0.36, 0.44), 0.28, 0.28,
                      boxstyle="round,pad=0.015,rounding_size=0.02",
                      linewidth=1.2, edgecolor="black", facecolor="#efe4ff")
ax.add_patch(box2)
ax.text(0.38, 0.69, "Symbolic level", fontsize=15, fontweight="bold")
ax.text(0.38, 0.65,
        r"$K\subset X,\quad C\subset X$" "\n"
        r"$K$ compact, $C$ closed," "\n"
        r"$K\cap C=\varnothing$" "\n\n"
        r"$\exists\,V\ni 0$ such that" "\n"
        r"$(K+V)\cap(C+V)=\varnothing$" "\n\n"
        r"$K+V=\bigcup_{x\in K}(x+V)$",
        ha="left", va="top", fontsize=13)

# Graphic box
box3 = FancyBboxPatch((0.67, 0.22), 0.28, 0.5,
                      boxstyle="round,pad=0.015,rounding_size=0.02",
                      linewidth=1.2, edgecolor="black", facecolor="#e9fff1")
ax.add_patch(box3)
ax.text(0.69, 0.69, "Graphic level", fontsize=15, fontweight="bold")

# Diagram inside graphic box
outer = Ellipse((0.81, 0.46), 0.22, 0.34, linewidth=2, edgecolor="black",
                facecolor="#ccebd6", alpha=0.8)
KplusV = Ellipse((0.77, 0.43), 0.13, 0.18, linewidth=2, edgecolor="black",
                 linestyle="--", facecolor="none")
Cset = Ellipse((0.85, 0.52), 0.09, 0.10, linewidth=1.5, edgecolor="black",
               facecolor="#bfc7d5")
Kset = Ellipse((0.75, 0.38), 0.08, 0.08, linewidth=1.5, edgecolor="black",
               facecolor="#ffd966")
vball = Circle((0.80, 0.27), 0.035, linewidth=1.2, edgecolor="black",
               facecolor="#d9e8ff")

for patch in [outer, KplusV, Cset, Kset, vball]:
    ax.add_patch(patch)

ax.text(0.77, 0.49, r"$K+V$", fontsize=14, fontweight="bold")
ax.text(0.75, 0.38, r"$K$", fontsize=14, ha="center", va="center")
ax.text(0.85, 0.52, r"$C$", fontsize=14, ha="center", va="center")
ax.text(0.80, 0.27, r"$V$", fontsize=12, ha="center", va="center")
ax.text(0.69, 0.61,
        "The same neighborhood $V$\n"
        "thickens both sets,\n"
        "but the enlarged sets remain\n"
        "disjoint.",
        ha="left", va="top", fontsize=11)

# Bottom note
note = FancyBboxPatch((0.05, 0.22), 0.59, 0.14,
                      boxstyle="round,pad=0.015,rounding_size=0.02",
                      linewidth=1.2, edgecolor="black", facecolor="#f3faff")
ax.add_patch(note)
ax.text(0.07, 0.31, "Interpretation", fontsize=15, fontweight="bold")
ax.text(0.07, 0.275,
        "Compactness gives a finite control of local separations.\n"
        "Closedness prevents limit points of $C$ from entering the buffer.\n"
        "Translation invariance lets one move neighborhoods from $0$ to points of $K$.",
        ha="left", va="top", fontsize=11)

out = "/mnt/data/separation_theorem_matplotlib.png"
plt.savefig(out, bbox_inches="tight")
plt.close(fig)

print(out)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyBboxPatch
import numpy as np

# ---------------------------
# Figure 1: R^2 nested neighborhoods
# ---------------------------
fig, ax = plt.subplots(figsize=(8, 8), dpi=180)
ax.set_aspect('equal')
ax.set_xlim(-4.5, 4.5)
ax.set_ylim(-4.5, 4.5)
ax.axis('off')

# Concentric circles
W = Circle((0, 0), 3.6, fill=False, linewidth=2.5)
UplusU = Circle((0, 0), 2.2, fill=False, linestyle='--', linewidth=2.2)
U = Circle((0, 0), 1.0, fill=False, linewidth=2.2)
for patch in [W, UplusU, U]:
    ax.add_patch(patch)

# Axes
ax.plot([-4.2, 4.2], [0, 0], linewidth=1.0)
ax.plot([0, 0], [-4.2, 4.2], linewidth=1.0)
ax.text(4.25, -0.18, r"$x$", fontsize=12)
ax.text(0.12, 4.15, r"$y$", fontsize=12)

# Labels
ax.plot(0, 0, 'o', markersize=4)
ax.text(0.12, 0.12, r"$0$", fontsize=12)
ax.text(2.75, 2.15, r"$W$", fontsize=16)
ax.text(1.55, 1.35, r"$U+U$", fontsize=16)
ax.text(0.65, 0.45, r"$U$", fontsize=16)

# Title and theorem statement
ax.text(0, 4.65, "Symmetric Neighborhood Lemma", ha='center', va='bottom',
        fontsize=18, fontweight='bold')
ax.text(0, 4.28, r"$U=-U \quad\mathrm{and}\quad U+U\subset W$",
        ha='center', va='bottom', fontsize=16)

# Annotation
ax.text(0, -4.7,
        "Geometric picture in $\\mathbb{R}^2$: choose a smaller symmetric neighborhood $U$\n"
        "so that adding two points of $U$ still lands inside $W$.",
        ha='center', va='top', fontsize=11)

out1 = "symmetric_neighborhood_r2.png"
plt.savefig(out1, bbox_inches="tight")
plt.close(fig)

# ---------------------------
# Figure 2: Real-line interval picture
# ---------------------------
fig, ax = plt.subplots(figsize=(10, 3.6), dpi=180)
ax.set_xlim(-4.5, 4.5)
ax.set_ylim(-1.2, 2.6)
ax.axis('off')

# baseline
ax.plot([-4.2, 4.2], [0, 0], linewidth=1.2)
ax.plot(0, 0, 'o', markersize=4)
ax.text(0.08, -0.22, r"$0$", fontsize=12)

# intervals
# W
ax.plot([-3.5, 3.5], [1.8, 1.8], linewidth=8, solid_capstyle='butt')
ax.text(3.68, 1.8, r"$W=(-\varepsilon,\varepsilon)$", va='center', fontsize=13)
# U+U
ax.plot([-2.0, 2.0], [1.0, 1.0], linewidth=8, linestyle='--', solid_capstyle='butt')
ax.text(2.18, 1.0, r"$U+U=(-2\delta,2\delta)$", va='center', fontsize=13)
# U
ax.plot([-1.0, 1.0], [0.2, 0.2], linewidth=8, solid_capstyle='butt')
ax.text(1.18, 0.2, r"$U=(-\delta,\delta)$", va='center', fontsize=13)

# endpoint guides
for x in [-3.5, 3.5, -2.0, 2.0, -1.0, 1.0]:
    ax.plot([x, x], [-0.08, 0.08], linewidth=1)

ax.text(0, 2.35, "Interval picture in $\\mathbb{R}$", ha='center',
        fontsize=18, fontweight='bold')
ax.text(0, -0.7,
        r"If $2\delta<\varepsilon$, then $U+U=(-2\delta,2\delta)\subset(-\varepsilon,\varepsilon)=W$.",
        ha='center', fontsize=13)

out2 = "symmetric_neighborhood_R.png"
plt.savefig(out2, bbox_inches="tight")
plt.close(fig)

# ---------------------------
# Figure 3: Product-space continuity picture
# ---------------------------
fig, ax = plt.subplots(figsize=(10, 4.8), dpi=180)
ax.set_xlim(0, 10)
ax.set_ylim(0, 5)
ax.axis('off')

# Left panel: X x X
left = FancyBboxPatch((0.4, 0.5), 4.1, 3.8, boxstyle="round,pad=0.02,rounding_size=0.04",
                      fill=False, linewidth=1.6)
ax.add_patch(left)
ax.text(2.45, 4.45, r"In $X\times X$", ha='center', fontsize=16, fontweight='bold')

# axes
ax.plot([0.9, 4.0], [2.35, 2.35], linewidth=1.0)
ax.plot([2.45, 2.45], [0.9, 3.9], linewidth=1.0)
ax.text(4.06, 2.18, r"$x$", fontsize=11)
ax.text(2.56, 3.92, r"$y$", fontsize=11)
ax.plot(2.45, 2.35, 'o', markersize=3)
ax.text(2.55, 2.18, r"$(0,0)$", fontsize=10)

# rectangle V1xV2
rect = Rectangle((1.55, 1.55), 1.8, 1.6, fill=False, linestyle='--', linewidth=2)
ax.add_patch(rect)
ax.text(2.45, 3.28, r"$V_1\times V_2$", ha='center', fontsize=13)

# Arrow
ax.annotate("", xy=(5.5, 2.45), xytext=(4.7, 2.45),
            arrowprops=dict(arrowstyle="->", linewidth=1.8))
ax.text(5.1, 2.7, r"$+$", ha='center', fontsize=18)

# Right panel: X
right = FancyBboxPatch((5.8, 0.5), 3.8, 3.8, boxstyle="round,pad=0.02,rounding_size=0.04",
                       fill=False, linewidth=1.6)
ax.add_patch(right)
ax.text(7.7, 4.45, r"In $X$", ha='center', fontsize=16, fontweight='bold')

# line and intervals
ax.plot([6.2, 9.2], [2.35, 2.35], linewidth=1.0)
ax.plot(7.7, 2.35, 'o', markersize=3)
ax.text(7.78, 2.12, r"$0$", fontsize=10)

# W big
ax.plot([6.35, 9.05], [3.2, 3.2], linewidth=7, solid_capstyle='butt')
ax.text(9.13, 3.2, r"$W$", va='center', fontsize=13)
# V1+V2 smaller
ax.plot([6.9, 8.5], [1.55, 1.55], linewidth=7, linestyle='--', solid_capstyle='butt')
ax.text(8.62, 1.55, r"$V_1+V_2$", va='center', fontsize=13)

ax.text(7.7, 0.15,
        r"Continuity of $+:X\times X\to X$ at $(0,0)$ gives"
        "\n"
        r"$V_1\times V_2 \mapsto V_1+V_2\subset W$.",
        ha='center', fontsize=12)

out3 = "sum_continuity_product_space.png"
plt.savefig(out3, bbox_inches="tight")
plt.close(fig)

print(out1)
print(out2)
print(out3)
```
