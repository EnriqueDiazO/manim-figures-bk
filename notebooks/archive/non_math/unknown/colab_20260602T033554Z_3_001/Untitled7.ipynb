```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, np.pi, 10000)
z = np.exp(1j * t)          # e^{it}
x, y = z.real, z.imag       # partes real e imaginaria

plt.figure()
plt.plot(x, y)
plt.gca().set_aspect("equal", adjustable="box")
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.xlabel("Re")
plt.ylabel("Im")
plt.title(r"Curve $e^{it}$, $t\in[0,\pi]$")
plt.show()
```
