```python
## Depuración de un bucle con `pdb` en Jupyter Notebook

### Código
import pdb

# Lista de números
numeros = [1, 2, 3, 0, 5]

# Bucle for
for num in numeros:
    pdb.set_trace()  # Detiene la ejecución en cada iteración, para continuar la ejecución escribe continue
    resultado = 10 / num
    print("El resultado es:", resultado)
```
