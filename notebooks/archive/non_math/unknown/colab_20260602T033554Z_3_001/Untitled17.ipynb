```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

# ---------- helpers ----------
def add_box(ax, x, y, w, h, text, edgecolor="#2e7d32", facecolor="#f8fff8",
            fontsize=16, lw=2.2):
    box = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.02,rounding_size=0.03",
        linewidth=lw, edgecolor=edgecolor, facecolor=facecolor
    )
    ax.add_patch(box)
    ax.text(
        x + w/2, y + h/2, text,
        ha="center", va="center", fontsize=fontsize
    )

def add_arrow(ax, x1, y1, x2, y2, color="#444444", lw=2.2, ms=18):
    arr = FancyArrowPatch(
        (x1, y1), (x2, y2),
        arrowstyle='-|>', mutation_scale=ms,
        linewidth=lw, color=color
    )
    ax.add_patch(arr)

# ---------- figure ----------
fig, ax = plt.subplots(figsize=(16, 8))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# Title
ax.text(
    0.5, 0.95,
    r"Construction of the Mellin convolution operator on $\mathbb{R}_{+}$",
    ha="center", va="center", fontsize=22, fontweight="bold"
)

# Top pipeline boxes
add_box(
    ax, 0.05, 0.62, 0.18, 0.16,
    r"Input function" "\n"
    r"$f \in L^{p}(\mathbb{R}_{+},w)$",
    edgecolor="#c62828", facecolor="#fff5f5", fontsize=17
)

add_box(
    ax, 0.29, 0.62, 0.20, 0.16,
    r"Isometric isomorphism" "\n"
    r"$E_{p,w}:L^{p}(\mathbb{R}_{+},w)\to L^{p}(\mathbb{R})$" "\n"
    r"$(E_{p,w}f)(x)=e^{x/p}w(e^x)f(e^x)$",
    edgecolor="#ad1457", facecolor="#fff5fb", fontsize=15
)

add_box(
    ax, 0.56, 0.62, 0.18, 0.16,
    r"Fourier convolution" "\n"
    r"$W^{0}(a)=\mathcal{F}^{-1}a\mathcal{F}$" "\n"
    r"acts on $L^{p}(\mathbb{R})$",
    edgecolor="#1565c0", facecolor="#f4f9ff", fontsize=16
)

add_box(
    ax, 0.80, 0.62, 0.15, 0.16,
    r"Inverse map" "\n"
    r"$E_{p,w}^{-1}$" "\n"
    r"back to $L^{p}(\mathbb{R}_{+},w)$",
    edgecolor="#2e7d32", facecolor="#f7fff7", fontsize=15
)

# arrows top
add_arrow(ax, 0.23, 0.70, 0.29, 0.70)
add_arrow(ax, 0.49, 0.70, 0.56, 0.70)
add_arrow(ax, 0.74, 0.70, 0.80, 0.70)

# labels above arrows
ax.text(0.26, 0.73, r"$E_{p,w}$", fontsize=18, ha="center", color="#ad1457")
ax.text(0.525, 0.73, r"$W^{0}(a)$", fontsize=18, ha="center", color="#1565c0")
ax.text(0.77, 0.73, r"$E_{p,w}^{-1}$", fontsize=18, ha="center", color="#2e7d32")

# Middle pipeline: text + arrows
y_mid = 0.47
ax.text(0.10, y_mid, r"$f$", fontsize=20, ha="center", va="center")
ax.text(0.35, y_mid, r"$E_{p,w}f$", fontsize=20, ha="center", va="center")
ax.text(0.62, y_mid, r"$W^{0}(a)\,E_{p,w}f$", fontsize=20, ha="center", va="center")
ax.text(0.88, y_mid, r"$E_{p,w}^{-1}W^{0}(a)E_{p,w}f$", fontsize=20, ha="center", va="center")

add_arrow(ax, 0.14, y_mid, 0.27, y_mid, color="#ad1457")
add_arrow(ax, 0.42, y_mid, 0.53, y_mid, color="#1565c0")
add_arrow(ax, 0.71, y_mid, 0.82, y_mid, color="#2e7d32")

ax.text(0.205, y_mid + 0.04, r"$E_{p,w}$", fontsize=15, ha="center", color="#ad1457")
ax.text(0.475, y_mid + 0.04, r"$W^{0}(a)$", fontsize=15, ha="center", color="#1565c0")
ax.text(0.765, y_mid + 0.04, r"$E_{p,w}^{-1}$", fontsize=15, ha="center", color="#2e7d32")

# Bottom result box
add_box(
    ax, 0.16, 0.18, 0.68, 0.16,
    r"Mellin convolution operator" "\n"
    r"$M^{0}(a)f := E_{p,w}^{-1}W^{0}(a)E_{p,w}f,"
    r"\qquad f\in L^{p}(\mathbb{R}_{+},w)$" "\n"
    r"$M^{0}(a)\in \mathcal{B}(L^{p}(\mathbb{R}_{+},w))$",
    edgecolor="#1b5e20", facecolor="#f3fff3", fontsize=18
)

# Down arrow to result
add_arrow(ax, 0.5, 0.41, 0.5, 0.34, color="#1b5e20")

# Footer interpretation
ax.text(
    0.5, 0.08,
    r"Interpretation: transport $f$ from $\mathbb{R}_{+}$ to $\mathbb{R}$ via $E_{p,w}$, "
    r"apply the Fourier convolution operator $W^{0}(a)$ on the line, "
    r"and return to $\mathbb{R}_{+}$ via $E_{p,w}^{-1}$.",
    ha="center", va="center", fontsize=15, color="#333333"
)

plt.tight_layout()
plt.show()
```
