```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Build a clean pedagogical schematic of an admissible curve
fig, ax = plt.subplots(figsize=(10, 6), dpi=200)

# Common vertex z
z = np.array([0.0, 0.0])

# Three Lyapunov-like smooth arcs meeting at z with different tangents
t = np.linspace(0, 1, 200)

# Gamma_1 directed away from z
g1 = np.vstack([2.6*t, 0.7*np.sin(0.7*np.pi*t) + 0.25*t]).T

# Gamma_2 directed towards z: draw curve from left endpoint to z
s = np.linspace(0, 1, 200)
g2 = np.vstack([-2.4*(1-s), 1.4*(1-s) - 0.35*np.sin(np.pi*s)]).T  # ends at z

# Gamma_3 directed away from z
g3 = np.vstack([1.35*t, -1.7*t + 0.25*np.sin(1.4*np.pi*t)]).T

def plot_arc(points, label, color, n_label, label_pos, arrow_index=120):
    ax.plot(points[:,0], points[:,1], lw=3, color=color, solid_capstyle="round")
    # direction arrow along the curve
    i = arrow_index
    ax.annotate(
        "",
        xy=(points[i+8,0], points[i+8,1]),
        xytext=(points[i-12,0], points[i-12,1]),
        arrowprops=dict(arrowstyle="->", lw=2.2, color=color)
    )
    ax.text(label_pos[0], label_pos[1], label, fontsize=13, color=color, weight="bold")
    ax.text(label_pos[0], label_pos[1]-0.22, n_label, fontsize=11, color=color)

plot_arc(g1, r"$\Gamma_1$", "#1f77b4", r"$n_1(z)=1$  (sale de $z$)", (1.65, 0.9), 110)
plot_arc(g2, r"$\Gamma_2$", "#d62728", r"$n_2(z)=-1$ (entra a $z$)", (-2.35, 1.35), 95)
plot_arc(g3, r"$\Gamma_3$", "#2ca02c", r"$n_3(z)=1$  (sale de $z$)", (1.0, -1.65), 100)

# Mark z and tangent rays
ax.scatter([0], [0], s=80, color="black", zorder=5)
ax.text(0.08, 0.08, r"$z$", fontsize=15, weight="bold")

# One-sided tangent lines at z (schematic)
r = np.linspace(0, 1.0, 20)
ax.plot(r*1.0, r*0.25, "--", color="#1f77b4", alpha=0.7, lw=1.5)
ax.plot(-r*0.85, r*0.55, "--", color="#d62728", alpha=0.7, lw=1.5)
ax.plot(r*0.65, -r*0.9, "--", color="#2ca02c", alpha=0.7, lw=1.5)
ax.text(-0.4, -0.42, "tangentes unilaterales\nno coinciden", fontsize=10,
        bbox=dict(boxstyle="round,pad=0.35", fc="white", ec="0.7"))

# Title and condition box
ax.set_title("Curva admisible: unión finita de arcos de Lyapunov con encuentros controlados",
             fontsize=14, pad=14)

condition_text = (
    "Condiciones en cada par $i\\ne j$:\n"
    r"$\Gamma_i\cap\Gamma_j=\varnothing$  o  $\{z\}$" "\n"
    "$z$ es extremo de ambos arcos\n"
    "las tangentes unilaterales en $z$ no coinciden"
)
ax.text(-3.0, -2.25, condition_text, fontsize=10.5,
        bbox=dict(boxstyle="round,pad=0.45", fc="#f8f8f8", ec="0.5"))

ax.set_aspect("equal", adjustable="box")
ax.set_xlim(-3.2, 3.0)
ax.set_ylim(-2.5, 2.1)
ax.axis("off")

out = Path("curva_admisible_esquema.png")
fig.savefig(out, bbox_inches="tight")
plt.close(fig)

out.as_posix()
```
