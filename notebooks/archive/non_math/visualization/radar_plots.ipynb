```python
import plotly.graph_objects as go

# Categorías
categories = [
    "Social Networks", "Email", "Internet Banking", "News Sports",
    "Search Engine", "View Shopping Sites", "Paying Online",
    "Buy Online", "Stream Music", "Online Gaming",
    "Navigation", "App connected to TV program",
    "Online Gaming (console)", "Photo Video",
    "Reading", "Listen Music", "Watch TV",
    "Listen Radio", "Sending Money", "Other"
]

# Datos
group_a = [85, 78, 60, 72, 80, 68, 55, 50, 45, 40, 35, 30, 28, 25, 32, 48, 52, 46, 38, 42]
group_b = [70, 82, 65, 60, 72, 75, 62, 58, 50, 55, 42, 38, 35, 30, 40, 55, 60, 50, 45, 48]

# 🔴 CERRAR POLÍGONO (clave)
categories_closed = categories + [categories[0]]
group_a_closed = group_a + [group_a[0]]
group_b_closed = group_b + [group_b[0]]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=group_a_closed,
    theta=categories_closed,
    fill='toself',
    name='Group A',
    line=dict(color='royalblue', width=2),
    line_shape='linear'   # 🔴 SIN CURVAS
))

fig.add_trace(go.Scatterpolar(
    r=group_b_closed,
    theta=categories_closed,
    fill='toself',
    name='Group B',
    line=dict(color='darkorange', width=2),
    line_shape='linear'   # 🔴 SIN CURVAS
))

fig.update_layout(
    title="Radar Chart (Flat Spider Web)",
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 100],
            gridcolor="lightgray"
        ),
        angularaxis=dict(
            gridcolor="lightgray"
        )
    ),
    showlegend=True
)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Datos
# -------------------------
labels = [
    "DM", "Hypertension", "Dyslipidemia", "Smoking", "Family\historySocial"
]

group_I = np.array([36, 91, 55, 18, 18])
group_II = 83, 87, 43, 13, 17]
group_II = 83, 87, 43, 13, 17]
values_a = np.array([85, 78, 60, 72, 80, 68, 55, 50, 45, 40])
values_b = np.array([70, 82, 65, 60, 72, 75, 62, 58, 50, 55])

max_value = 100
levels = 5 # anillos de la telaraña

# -------------------------
# Geometría
# -------------------------
N = len(labels)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)
angles_closed = np.concatenate([angles, [angles[0]]])

values_a = np.concatenate([values_a, [values_a[0]]])
values_b = np.concatenate([values_b, [values_b[0]]])

# -------------------------
# Figura
# -------------------------
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect("equal")
ax.axis("off")

# -------------------------
# Telaraña (poligonal)
# -------------------------
for level in range(1, levels + 1):
    radius = max_value * level / levels
    x = radius * np.cos(angles_closed)
    y = radius * np.sin(angles_closed)
    ax.plot(x, y, color="gray", linewidth=0.5)

# Ejes radiales
for angle in angles:
    ax.plot(
        [0, max_value * np.cos(angle)],
        [0, max_value * np.sin(angle)],
        color="gray",
        linewidth=0.5
    )

# -------------------------
# Datos
# -------------------------
ax.plot(
    values_a * np.cos(angles_closed),
    values_a * np.sin(angles_closed),
    color="#FFA500",
    linewidth=2,
    label="Group A"
)

ax.plot(
    values_b * np.cos(angles_closed),
    values_b * np.sin(angles_closed),
    color="#FF1493",
    linewidth=2,
    label="Group B"
)

# -------------------------
# Etiquetas
# -------------------------
for angle, label in zip(angles, labels):
    x = (max_value + 10) * np.cos(angle)
    y = (max_value + 10) * np.sin(angle)
    ax.text(x, y, label, ha="center", va="center", fontsize=9)

ax.legend(loc="upper right")
plt.show()
```
