```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, FancyBboxPatch

# ------------------------------------------------------------
# Diagram: Operational role of the Fourier symbol a
# W^0(a) = F^{-1} a F
# ------------------------------------------------------------

def add_box(ax, xy, text, width=2.25, height=0.85, fontsize=22):
    """
    Draws a rounded box with centered LaTeX text.
    """
    x, y = xy
    box = FancyBboxPatch(
        (x - width / 2, y - height / 2),
        width,
        height,
        boxstyle="round,pad=0.12,rounding_size=0.08",
        linewidth=1.4,
        edgecolor="black",
        facecolor="white"
    )
    ax.add_patch(box)
    ax.text(
        x, y, text,
        ha="center",
        va="center",
        fontsize=fontsize
    )


def add_arrow(ax, start, end, label=None, label_offset=(0, 0), fontsize=16):
    """
    Draws an arrow from start to end and optionally adds a label.
    """
    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=18,
        linewidth=1.6,
        color="black"
    )
    ax.add_patch(arrow)

    if label is not None:
        lx = (start[0] + end[0]) / 2 + label_offset[0]
        ly = (start[1] + end[1]) / 2 + label_offset[1]
        ax.text(
            lx, ly, label,
            ha="center",
            va="center",
            fontsize=fontsize
        )


# Figure setup
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis("off")

# Node positions
pos_f = (2.2, 4.2)
pos_Ff = (7.8, 4.2)
pos_aFf = (7.8, 1.8)
pos_Waf = (2.2, 1.8)

# Title
ax.text(
    5, 5.55,
    r"Operational symbol of a Fourier convolution operator",
    ha="center",
    va="center",
    fontsize=20
)

# Main formula
ax.text(
    5, 5.05,
    r"$W^0(a)=\mathcal{F}^{-1}a\mathcal{F}$",
    ha="center",
    va="center",
    fontsize=24
)

# Boxes
add_box(ax, pos_f, r"$f$")
add_box(ax, pos_Ff, r"$\mathcal{F}f$")
add_box(ax, pos_aFf, r"$a\,\mathcal{F}f$")
add_box(ax, pos_Waf, r"$W^0(a)f$")

# Arrows
add_arrow(
    ax,
    (3.35, 4.2),
    (6.65, 4.2),
    label=r"$\mathcal{F}$",
    label_offset=(0, 0.32),
    fontsize=20
)

add_arrow(
    ax,
    (7.8, 3.75),
    (7.8, 2.25),
    label=r"multiply by $a$",
    label_offset=(1.05, 0),
    fontsize=15
)

add_arrow(
    ax,
    (6.65, 1.8),
    (3.35, 1.8),
    label=r"$\mathcal{F}^{-1}$",
    label_offset=(0, -0.35),
    fontsize=20
)

# Explanatory footer
ax.text(
    5, 0.65,
    r"The symbol $a$ acts in the Fourier variable.",
    ha="center",
    va="center",
    fontsize=16
)

# Save files
plt.tight_layout()
plt.savefig("fourier_symbol_diagram.png", dpi=300, bbox_inches="tight")
plt.savefig("fourier_symbol_diagram.pdf", bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
from matplotlib import rcParams

rcParams["mathtext.fontset"] = "cm"
rcParams["font.family"] = "DejaVu Sans"

fig, ax = plt.subplots(figsize=(17, 5.2), dpi=220)
ax.set_xlim(0, 17)
ax.set_ylim(0, 5.2)
ax.axis("off")

colors = {
    "input_fill": "#FFF7CC",
    "input_edge": "#DAA520",
    "fourier_fill": "#E2F1FF",
    "fourier_edge": "#2E73D2",
    "symbol_fill": "#FFEAD9",
    "symbol_edge": "#DA5F2D",
    "inverse_fill": "#F0E8FF",
    "inverse_edge": "#7855C8",
    "output_fill": "#E1F6DC",
    "output_edge": "#469646",
    "group_fill": "#F5F8FF",
    "group_edge": "#87A5DC",
    "text_dark": "#111111",
}

def add_box(center, text, fill, edge, width=2.35, height=0.82, fontsize=18):
    x, y = center
    box = FancyBboxPatch(
        (x - width/2, y - height/2),
        width, height,
        boxstyle="round,pad=0.17,rounding_size=0.12",
        linewidth=1.7,
        edgecolor=edge,
        facecolor=fill,
        zorder=4
    )
    ax.add_patch(box)
    ax.text(x, y, text, ha="center", va="center", fontsize=fontsize, color=colors["text_dark"], zorder=5)

def add_arrow(start, end, color, label, note, label_y=0.34, note_y=-0.38):
    arrow = FancyArrowPatch(
        start, end,
        arrowstyle="-|>",
        mutation_scale=18,
        linewidth=2.0,
        color=color,
        zorder=3
    )
    ax.add_patch(arrow)
    mx, my = (start[0]+end[0])/2, (start[1]+end[1])/2
    ax.text(mx, my+label_y, label, ha="center", va="center", fontsize=20, color=color, zorder=5)
    ax.text(mx, my+note_y, note, ha="center", va="center", fontsize=10.5, style="italic", color=color, zorder=5)

y = 2.55
xpos = [1.55, 4.65, 7.75, 10.95, 14.15]

group = FancyBboxPatch(
    (3.15, 1.48),
    6.15, 2.14,
    boxstyle="round,pad=0.24,rounding_size=0.18",
    linewidth=1.4,
    linestyle="--",
    edgecolor=colors["group_edge"],
    facecolor=colors["group_fill"],
    alpha=0.95,
    zorder=1
)
ax.add_patch(group)

ax.text(8.5, 4.72, r"$W^0(a)=\mathcal{F}^{-1}a\mathcal{F}$",
        ha="center", va="center", fontsize=28, color=colors["text_dark"])
ax.text(8.5, 4.24, "Operational role of the Fourier symbol",
        ha="center", va="center", fontsize=18, color="#333333")

ax.text(xpos[0], 3.85, "original side", ha="center", va="center",
        fontsize=11, style="italic", color=colors["input_edge"])
ax.text((xpos[1]+xpos[2])/2, 3.85, "Fourier side", ha="center", va="center",
        fontsize=11, style="italic", color=colors["fourier_edge"])
ax.text(xpos[4], 3.85, "output side", ha="center", va="center",
        fontsize=11, style="italic", color=colors["output_edge"])

add_box((xpos[0], y), r"$f\in L^p(\mathbb{R},w)$",
        colors["input_fill"], colors["input_edge"], width=2.55, fontsize=17)
add_box((xpos[1], y), r"$\mathcal{F}f$",
        colors["fourier_fill"], colors["fourier_edge"], width=2.15, fontsize=20)
add_box((xpos[2], y), r"$a\,(\mathcal{F}f)$",
        colors["symbol_fill"], colors["symbol_edge"], width=2.35, fontsize=20)
add_box((xpos[3], y), r"$\mathcal{F}^{-1}(a\,\mathcal{F}f)$",
        colors["inverse_fill"], colors["inverse_edge"], width=2.75, fontsize=18)
add_box((xpos[4], y), r"$W^0(a)f$",
        colors["output_fill"], colors["output_edge"], width=2.15, fontsize=20)

add_arrow((2.85, y), (3.55, y), colors["fourier_edge"], r"$\mathcal{F}$", "Fourier transform")
add_arrow((5.75, y), (6.55, y), colors["symbol_edge"], r"$a$", "multiply by symbol")
add_arrow((8.95, y), (9.65, y), colors["inverse_edge"], r"$\mathcal{F}^{-1}$", "inverse transform")
add_arrow((12.35, y), (13.05, y), colors["output_edge"], r"$=$", r"definition of $W^0(a)$", label_y=0.28)

note_box = FancyBboxPatch(
    (4.0, 0.48), 9.0, 0.72,
    boxstyle="round,pad=0.18,rounding_size=0.12",
    linewidth=1.1,
    edgecolor="#BBBBBB",
    facecolor="#FBFBFB",
    zorder=2
)
ax.add_patch(note_box)
ax.text(8.5, 0.84,
        r"The symbol $a$ acts after transporting $f$ to the Fourier variable.",
        ha="center", va="center", fontsize=14, color="#222222", zorder=5)

png_path = "fourier_symbol_diagram_colored.png"
pdf_path = "fourier_symbol_diagram_colored.pdf"
svg_path = "fourier_symbol_diagram_colored.svg"

plt.savefig(png_path, bbox_inches="tight", facecolor="white")
plt.savefig(pdf_path, bbox_inches="tight", facecolor="white")
plt.savefig(svg_path, bbox_inches="tight", facecolor="white")
plt.show()
print(png_path)
print(pdf_path)
print(svg_path)
```
