```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Circle
from matplotlib.lines import Line2D
import numpy as np

fig, ax = plt.subplots(figsize=(13, 8))
ax.set_xlim(0, 13)
ax.set_ylim(0, 9)
ax.axis("off")

def box(x, y, w, h, text, edge="black", face="#f8f8f8", size=14, weight="normal"):
    patch = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.03,rounding_size=0.08",
        linewidth=1.5, edgecolor=edge, facecolor=face
    )
    ax.add_patch(patch)
    ax.text(x + w/2, y + h/2, text, ha="center", va="center",
            fontsize=size, fontweight=weight)
    return patch

def arrow(x1, y1, x2, y2, text=None, color="black", ls="-", curve=0.0):
    arr = FancyArrowPatch(
        (x1, y1), (x2, y2),
        arrowstyle="-|>", mutation_scale=14,
        linewidth=1.3, color=color, linestyle=ls,
        connectionstyle=f"arc3,rad={curve}"
    )
    ax.add_patch(arr)
    if text:
        ax.text((x1+x2)/2, (y1+y2)/2 + 0.25, text,
                fontsize=11, ha="center", va="center", color=color)

# ------------------------------------------------------------
# Left side: half-line and algebra on R_+
# ------------------------------------------------------------
box(0.8, 7.4, 3.6, 0.9, r"$SO^\diamond(\mathbb{R}_+)$",
    edge="#3366cc", face="#dfeaff", size=18, weight="bold")

# Draw R_+
ax.add_line(Line2D([0.9, 4.0], [4.8, 4.8], linewidth=2.0, color="black"))
ax.add_patch(Circle((1.0, 4.8), 0.06, color="black"))
ax.text(0.9, 4.35, r"$0$", fontsize=13)
ax.text(2.2, 5.15, r"$\mathbb{R}_+=(0,\infty)$", fontsize=14)
ax.text(3.85, 5.05, r"$x$", fontsize=13)

# ------------------------------------------------------------
# Middle: parametrizations
# ------------------------------------------------------------
box(4.9, 6.25, 3.2, 1.0, r"$\phi_k:\mathbb{R}_+\to \Gamma_k$",
    edge="#8b5a2b", face="#fff1db", size=16, weight="bold")

ax.text(6.5, 5.55, r"transport by parametrization", fontsize=12, ha="center")

ax.text(6.5, 4.95, r"$\phi_k(x)=\dfrac{1}{x-ic_k},\quad k=1,\dots,s$",
        fontsize=13, ha="center", color="#1f4fa3")
ax.text(6.5, 4.35, r"$\phi_k(x)=\dfrac{e^{i\beta_k}}{x},\quad k=s+1,\dots,m$",
        fontsize=13, ha="center", color="#b00035")

# ------------------------------------------------------------
# Right side: schematic star-like curve Gamma
# ------------------------------------------------------------
box(9.0, 7.4, 3.0, 0.9, r"$SO^\diamond(\Gamma)$",
    edge="#7a3db8", face="#f2e9ff", size=18, weight="bold")

# center point
cx, cy = 10.5, 4.8
ax.add_patch(Circle((cx, cy), 0.07, color="black"))
ax.text(cx + 0.12, cy - 0.28, r"$0$", fontsize=12)

# cusp-like arcs (blue)
t = np.linspace(0.15, 1.0, 150)
x1 = cx - 1.8 * t
y1 = cy + 1.3 * t**2
ax.plot(x1, y1, color="#3366cc", linewidth=2)
ax.text(cx - 1.7, cy + 1.45, r"$\Gamma_1$", fontsize=12, color="#3366cc")

x2 = cx - 1.7 * t
y2 = cy + 0.6 * t**2
ax.plot(x2, y2, color="#3366cc", linewidth=2)
ax.text(cx - 1.55, cy + 0.8, r"$\Gamma_s$", fontsize=12, color="#3366cc")

# ray-like arcs (red)
angles = [-0.4, -0.9, -1.25]
labels = [r"$\Gamma_{s+1}$", r"$\Gamma_{s+2}$", r"$\Gamma_m$"]
for ang, lab, rr in zip(angles, labels, [1.8, 1.65, 1.55]):
    x_end = cx + rr * np.cos(ang)
    y_end = cy + rr * np.sin(ang)
    ax.plot([cx, x_end], [cy, y_end], color="#b00035", linewidth=2)
    ax.text(x_end + 0.05, y_end - 0.05, lab, fontsize=11, color="#b00035")

ax.text(10.5, 5.95, r"$\Gamma=\bigcup_{k=1}^m \Gamma_k$", fontsize=14, ha="center")

# ------------------------------------------------------------
# Arrows left -> middle -> right
# ------------------------------------------------------------
arrow(4.0, 4.8, 4.9, 6.75, text=r"$x \mapsto \phi_k(x)$", color="#8b5a2b", curve=0.15)
arrow(8.1, 6.75, 9.0, 4.95, text=r"pullback / transfer", color="#8b5a2b", curve=-0.15)

# ------------------------------------------------------------
# Bottom equivalence statement
# ------------------------------------------------------------
ax.text(6.5, 2.1,
        r"$f\in SO^\diamond(\Gamma)\quad \Longleftrightarrow \quad f_k\circ\phi_k \in SO^\diamond(\mathbb{R}_+)\ \ \text{for each }k$",
        fontsize=14, ha="center")

# Title
ax.text(0.5, 8.75,
        r"Figure 4. Transfer of slowly oscillating structure from $\mathbb{R}_+$ to the arcs of $\Gamma$",
        fontsize=18, fontweight="bold", ha="left")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Ellipse, Circle
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(13, 8))
ax.set_xlim(0, 13)
ax.set_ylim(0, 9)
ax.axis("off")

def box(x, y, w, h, text, edge="black", face="#f8f8f8", size=14, weight="normal"):
    patch = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.03,rounding_size=0.08",
        linewidth=1.5, edgecolor=edge, facecolor=face
    )
    ax.add_patch(patch)
    ax.text(x + w/2, y + h/2, text, ha="center", va="center",
            fontsize=size, fontweight=weight)
    return patch

# ------------------------------------------------------------
# Top: maximal ideal space
# ------------------------------------------------------------
box(3.0, 7.7, 7.0, 0.9, r"$M(SO^\diamond)$",
    edge="#7a3db8", face="#f2e9ff", size=19, weight="bold")

# ------------------------------------------------------------
# Projection arrow
# ------------------------------------------------------------
proj = FancyArrowPatch(
    (6.5, 7.7), (6.5, 6.6),
    arrowstyle="-|>", mutation_scale=15,
    linewidth=1.4, color="black"
)
ax.add_patch(proj)
ax.text(6.85, 7.05, r"$\xi \mapsto \xi|_{C(\dot{\mathbb{R}})}=\delta_\lambda$",
        fontsize=12, ha="left", va="center")

# ------------------------------------------------------------
# Base space: R dot
# ------------------------------------------------------------
ax.add_line(Line2D([1.2, 11.8], [2.0, 2.0], linewidth=2.0, color="black"))
ax.text(6.5, 1.2, r"$\dot{\mathbb{R}}=\mathbb{R}\cup\{\infty\}$", fontsize=15, ha="center")

# marked points on base
base_points = [
    (2.0, r"$\lambda_1$"),
    (4.3, r"$\lambda_2$"),
    (6.5, r"$\lambda$"),
    (8.7, r"$\lambda_3$"),
    (11.0, r"$\infty$")
]

for x, label in base_points:
    ax.add_patch(Circle((x, 2.0), 0.07, color="black"))
    ax.text(x, 1.65, label, fontsize=12, ha="center")

# ------------------------------------------------------------
# Fibers above points
# ------------------------------------------------------------
fiber_specs = [
    (2.0, 4.2, 0.75, 1.8, r"$M_{\lambda_1}(SO^\diamond)$"),
    (4.3, 4.4, 0.75, 1.7, r"$M_{\lambda_2}(SO^\diamond)$"),
    (6.5, 4.5, 0.85, 2.0, r"$M_{\lambda}(SO^\diamond)$"),
    (8.7, 4.3, 0.75, 1.7, r"$M_{\lambda_3}(SO^\diamond)$"),
    (11.0, 4.4, 0.95, 2.1, r"$M_{\infty}(SO^\diamond)$"),
]

for x, y, w, h, label in fiber_specs:
    # vertical guide
    ax.add_line(Line2D([x, x], [2.1, y - h/2], linewidth=1.1, color="#888888", linestyle="--"))
    # fiber cloud
    e = Ellipse((x, y), width=w, height=h,
                linewidth=1.5, edgecolor="#3366cc", facecolor="#dfeaff")
    ax.add_patch(e)
    ax.text(x, y, label, fontsize=11, ha="center", va="center", rotation=90)

# ------------------------------------------------------------
# Explanatory formulas
# ------------------------------------------------------------
ax.text(6.5, 6.0,
        r"$M(SO^\diamond)=\bigcup_{\lambda\in\dot{\mathbb{R}}} M_\lambda(SO^\diamond)$",
        fontsize=16, ha="center", color="#7a3db8")

ax.text(6.5, 5.4,
        r"$M_\lambda(SO^\diamond)=\{\xi\in M(SO^\diamond):\xi|_{C(\dot{\mathbb{R}})}=\delta_\lambda\}$",
        fontsize=13, ha="center")

ax.text(6.5, 0.45,
        "Schematic fiber decomposition of the maximal ideal space over the base space",
        fontsize=12, ha="center")

# ------------------------------------------------------------
# Title
# ------------------------------------------------------------
ax.text(0.5, 8.85,
        r"Figure 5. Fiber decomposition of the maximal ideal space of $SO^\diamond$",
        fontsize=18, fontweight="bold", ha="left")

plt.tight_layout()
plt.show()
```
