```python
def SigMov(r,c,princess):
    """
    Determina el siguiente movimiento para acercarse a la princesa.

    Parámetros:
    r (int): Fila actual del personaje.
    c (int): Columna actual del personaje.
    grid (list): Lista de cadenas que representa la cuadrícula.

    Retorna:
    str: El siguiente movimiento ('UP', 'DOWN', 'LEFT', 'RIGHT').
    """
    # Verifica si la princesa está arriba del personaje
    if princess[0] < r:
        move = "UP"
    # Verifica si la princesa está abajo del personaje
    elif princess[0] > r:
        move = "DOWN"
    # Verifica si la princesa está a la izquierda del personaje
    elif princess[1] < c:
        move = "LEFT"
    # Verifica si la princesa está a la derecha del personaje
    elif princess[1] > c:
        move = "RIGHT"
    else:
        move = "STAY"  # Caso en el que el personaje ya está en la posición de la princesa

    return move

def ImprGridMov(n, r, c, grid, move):
    """
    Muestra la cuadrícula con el movimiento actualizado del personaje.

    Parámetros:
    n (int): Tamaño de la cuadrícula (n x n).
    r (int): Fila actual del personaje.
    c (int): Columna actual del personaje.
    grid (list): Lista de cadenas que representa la cuadrícula.
    move (str): Movimiento que se va a realizar ('UP', 'DOWN', 'LEFT', 'RIGHT').
    """
    # Mueve el personaje según la dirección
    if move == "UP":
        r = r - 1
    elif move == "DOWN":
        r = r + 1
    elif move == "LEFT":
        c = c - 1
    elif move == "RIGHT":
        c = c + 1

    # Muestra la cuadrícula con la posición actualizada del personaje
    for i in range(n):
        # Convertir la fila actual en una lista para manipular los caracteres
        row = list(grid[i])
        # Reemplazar cualquier carácter 'm' con '.' antes de marcar la nueva posición
        if 'm' in row:
            row[row.index('m')] = '-'
        # Marca la nueva posición del personaje con 'm'
        if i == r:
            row[c] = 'm'
        # Convertir la lista de vuelta a una cadena y imprimir
        print(''.join(row))


# Lee el tamaño de la cuadrícula
n = int(input())
# Lee la posición del personaje (fila y columna)
x, y = [int(i) for i in input().strip().split()]
grid = []

# Lee la cuadrícula
for i in range(n):
  row = input().strip()
  if 'p' in row:
    # Encuentra la posición de la princesa
    princess = (i, row.index('p'))
  grid.append(row)

# Calcula el siguiente movimiento
move = SigMov(x, y, princess)

# Imprime el siguiente movimiento
print(move)

# Muestra la cuadrícula con el movimiento actualizado
ImprGridMov(n, x, y, grid, move)
```
