```python
# Importo las funciones raiz (sqrt), parte entera (ceil), factorial
from math import sqrt, ceil, factorial
# Defino un try-except para verificar que la entrada es un entero
try:
  # Recibo el entero del usuario
  x = int(input("Ingresa un número: "))
  # Defino una variable flag para comprobar si encontre el n
  flag = 0
  # Establezco un for desde 1 hasta la parte entera de raiz de x
  for n in range(1,ceil(sqrt(x))):
    # Compruebo si n! == x
    if factorial(n) == x:
      # Si lo encuentro, lo imprimo
      print(f"{n}!={x}")
      # Cambio flag a 1
      flag = 1
      # Cierro el for
      break
  # Si no encontré el n, imprimo Ninguno. Siempre y cuando n esté entre [1,parte_entera(raiz (x))]
  if flag==0:
    print("Ninguno")
except:
  print("No es un número")
  exit()
```
