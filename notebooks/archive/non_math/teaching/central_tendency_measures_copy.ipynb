```r
rm(list=ls())
###########################
# Medidas de tendencia central
###########################
###   Media aritmetica
# Datos de las calificaciones
calificaciones <- c(85, 90, 78, 92, 88, 76, 95, 89, 84, 91)

# Cálculo de la media aritmética
media_aritmetica <- mean(calificaciones)

# Mostrar la media aritmética
media_aritmetica
```

```r
rm(list=ls())
## Media geometrica
# Datos de los rendimientos
rendimientos <- c(1.05, 1.10, 1.03, 1.08, 1.02)

# Cálculo de la media geométrica
media_geometrica <- (prod(rendimientos))^(1/length(rendimientos))

# Mostrar la media geométrica
media_geometrica
```

```r
# Media armonica
rm(list=ls())
# Datos de las velocidades
velocidades <- c(60, 80, 100, 90, 70)

# Cálculo de la media armónica
media_armonica <- length(velocidades) / sum(1 / velocidades)

# Mostrar la media armónica
media_armonica
```

```r
# Mediana
rm(list=ls())

# Definimos los datos
data <- c(7, 2, 5, 1, 3, 6, 4)

# Calculamos la mediana
mediana <- median(data)

# Mostramos el resultado
print(mediana)

## Total de datos par
# Definimos un conjunto de datos con un número par de elementos
datos_par <- c(10, 15, 20, 25, 30, 35)

# Calculamos la mediana
mediana <- median(datos_par)

# Mostramos el resultado
print(mediana)
```

```r
# Moda
# Función para calcular la moda
calcular_moda <- function(x) {
  # Usamos la función table() para contar la frecuencia de cada valor
  frec <- table(x)
  # Devolvemos el valor o valores con la frecuencia máxima
  moda <- as.numeric(names(frec)[frec == max(frec)])
  return(moda)
}
# Definimos el conjunto de datos
datos <- c(4, 2, 2, 8, 5, 6, 4, 3, 4, 6)

# Calculamos la moda
moda <- calcular_moda(datos)

# Mostramos el resultado
print(moda)

### Multiples valores
datos <- c(1, 2, 2, 3, 3, 4, 5)
moda <- calcular_moda(datos)
print(moda)
```
