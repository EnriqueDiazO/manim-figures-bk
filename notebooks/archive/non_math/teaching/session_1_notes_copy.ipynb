# Tema 1. Introducción

## 1.1. Lenguaje de Programación Python.

Python es un lenguaje de programación:

- **De alto nivel:** Significa que está diseñado para ser fácil de leer y escribir para los humanos. Su sintaxis es clara y concisa, lo que facilita la comprensión del código. Esto permite a los programadores enfocarse en la lógica del problema en lugar de preocuparse por detalles de bajo nivel.

- **Interpretado:** En lugar de compilar el código a un lenguaje de máquina antes de ejecutarlo, como lo hacen los lenguajes compilados, Python utiliza un intérprete que ejecuta el código directamente. Esto significa que puedes escribir y probar código Python rápidamente sin necesidad de esperar por un proceso de compilación.

- **De propósito general:** Python se puede utilizar para una amplia variedad de tareas y aplicaciones. Desde el desarrollo de sitios web y aplicaciones de escritorio hasta el análisis de datos, la automatización de tareas y la inteligencia artificial, Python es una herramienta versátil que se adapta a numerosos contextos y proyectos.




## 1.2. Áreas de aplicación de Python.

Python es un lenguaje de programación versátil que se utiliza en una amplia variedad de aplicaciones, incluyendo:

1. **Desarrollo web:** Frameworks como Django y Flask permiten desarrollar aplicaciones web robustas y escalables.

2. **Aplicaciones de escritorio:** Con bibliotecas como Tkinter, PyQt y wxPython, Python se utiliza para crear interfaces de usuario (UI) intuitivas y funcionales.

3. **Análisis de datos:** Bibliotecas como Pandas, NumPy y Matplotlib hacen que Python sea una opción popular para análisis y visualización de datos.

4. **Inteligencia artificial y machine learning:** Con herramientas como TensorFlow, PyTorch y scikit-learn, Python se utiliza para desarrollar modelos de machine learning y aplicaciones de inteligencia artificial.

5. **Automatización de tareas:** Python es ampliamente utilizado para automatizar tareas repetitivas y procesos de flujo de trabajo en sistemas informáticos, administración de servidores, administración de redes, entre otros.

6. **Desarrollo de juegos:** A través de bibliotecas como Pygame y Pyglet, Python se utiliza en el desarrollo de juegos tanto para entretenimiento como para educación.

7. **Desarrollo de aplicaciones móviles:** Frameworks como Kivy permiten desarrollar aplicaciones móviles multiplataforma utilizando Python.

8. **Procesamiento de imágenes y visión por computadora:** Con bibliotecas como OpenCV y scikit-image, Python se utiliza para el procesamiento de imágenes, reconocimiento de objetos y visión por computadora.

## 1.3 Python 2 vs Python 3.

## 2.1. PyPI Package Manager

PyPI (Python Package Index) es el repositorio oficial de paquetes de software para Python. Pip (Python Install Package) es una herramienta utilizada para instalar y administrar paquetes desde PyPI.
## 2.2. Listar todos los paquetes instalados usando pip

Para listar todos los paquetes instalados en tu sistema, puedes ejecutar el siguiente comando en tu terminal:

````python
#Listado de Paquetes Instalados
!pip list
````

## 2.3. Actualizar Paquetes

Para actualizar un paquete instalado a la última versión disponible, puedes usar el siguiente comando:

pip install --upgrade <nombre_del_paquete>

## 2.4. Desinstalar Paquetes

Para desinstalar un paquete, puedes utilizar el siguiente comando:

pip uninstall <nombre_del_paquete>


## 2.5. Actualizar todos los paquetes obsoletos en Linux
Para visualizar todos los paquetes obsoletos en un sistema Linux, puedes ejecutar el siguiente comando:

````python
!pip list --outdated
````

## 2.7. Crear un archivo requirements.txt de todos los paquetes en el sistema

Para crear un archivo requirements.txt que contenga todos los paquetes instalados en tu sistema, puedes ejecutar el siguiente comando:

````python
pip freeze > requirements.txt
````

Para instalar un archivo requirements.txt que contiene una lista de dependencias de Python, simplemente puedes usar el siguiente comando:

````python
pip install -r requirements.txt
````

Para instalar un paquete desde un archivo de rueda (wheel), puedes utilizar el siguiente comando:

````python
pip install nombre_del_archivo.whl
````

## Tema 3: Módulo, Paquetes y Namespaces en Python

En Python, los módulos, paquetes y namespaces son componentes esenciales para organizar y estructurar programas de manera efectiva. A continuación, exploraremos cada uno de ellos en detalle:
Módulos en Python:

### Módulos
Los módulos son archivos de Python que encapsulan código reusable. Pueden contener variables, funciones y clases que se pueden utilizar en otros programas. Para emplear un módulo en un programa, es necesario importarlo primero.

````python
# Usando el módulo en otro archivo
import mimodulo

mimodulo.saludar("Juan")
````

### Paquetes en Python:

Los paquetes son directorios que agrupan módulos relacionados. Para ser reconocidos como paquetes, deben contener un archivo especial llamado __init__.py. Esto ayuda a organizar y estructurar el código de manera más jerárquica.

mi_paquete/
│   __init__.py
│   modulo1.py
│   modulo2.py


## Namespaces en Python:

Un namespace es un espacio que contiene nombres de variables, funciones, clases, etc. Ayuda a evitar conflictos de nombres y organizar el código de manera más clara.

Ejemplo de namespaces en Python:

````python
class MiClase:
    def __init__(self, nombre):  # Agregamos el parámetro nombre al constructor
        self.nombre = nombre     # Nuevo atributo de la clase
        self.valor = 5           # Atributo existente


# Creamos un objeto de la clase MiClase con el nombre "My_var"
objeto = MiClase("My_var")

# Accedemos a los atributos del objeto e imprimimos sus valores
print(objeto.nombre)
print(objeto.valor)
````
