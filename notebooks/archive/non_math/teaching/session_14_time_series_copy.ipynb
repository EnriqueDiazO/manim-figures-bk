## Técnicas de Análisis de Series Temporales

## - 1. **Autorregresión (AR): Autoregression (AR)**

Modelo que utiliza las dependencias entre una observación y un cierto número de observaciones anteriores (conocido como retraso o *lag*).
Su fórmula general es
$$
X_{t} = c + \phi_{1}X_{t-1} + \phi X_{t-2} + ... + \phi X_{t-p} + \epsilon_{t}
$$
Donde:
- $X_{t}$ es el valor actual
- $c$ es una constante
- $\phi_{i}$ son los coeficientes
- $\epsilon_{t}$ es el error



## - 2. **Promedio Móvil (MA): Moving Average (MA)**
Modelo que utiliza las dependencias entre una observación y un cierto número de errores anteriores.
Su fórmula general es
$$
X_{t} = \mu + \epsilon_{t} + \theta_{1}\epsilon_{t-1} + \theta_{2}\epsilon_{t-2} +...+\theta_{q}\epsilon_{t-q}
$$
Donde:
- $X_{t}$ es el valor actual
- $\mu$ es la media de la serie
- $\theta_{i}$ son los coeficientes
- $\epsilon_{t}$ es el error

## - 3. **Media Móvil Autorregresiva (ARMA): Autoregressive Moving Average (ARMA)**
Combina los modelos AR y MA.

Fórmula general:

$$
X_{t} = c + \phi_{1}X_{t-1} + \phi X_{t-2} + ... + \phi X_{t-p} + \epsilon_{t} + + \theta_{1}\epsilon_{t-1} + \theta_{2}\epsilon_{t-2} +...+\theta_{q}\epsilon_{t-q}
$$


## - 4. **Media Móvil Integrada Autorregresiva (ARIMA): Autoregressive Integrated Moving Average (ARIMA)**

> El modelo ARIMA (Autoregressive Integrated Moving Average) es una extensión del modelo ARMA (Autoregressive Moving Average) que permite trabajar con series de tiempo no estacionarias. Esta técnica combina componentes de autorregresión, promedio móvil e integración para modelar y predecir series de tiempo que muestran tendencias y patrones de variación.

>**Componentes del modelo ARIMA**:

> Autorregresión (AR):
        El componente autorregresivo modela la relación entre una observación y un número de observaciones pasadas. Se denota como $p$.
        Un modelo AR(p) utiliza pp observaciones pasadas para predecir la siguiente observación.

> Media Móvil (MA):
        El componente de media móvil modela la relación entre una observación y un error residual de un modelo de autorregresión aplicado a observaciones pasadas. Se denota como $q$.
        Un modelo MA(q) utiliza qq errores pasados para predecir la siguiente observación.

> Integración (I):
        El componente de integración se refiere al número de diferencias necesarias para hacer que una serie de tiempo sea estacionaria, es decir, para eliminar cualquier tendencia o patrón sistemático. Se denota como $d$.

### Nota: Una serie de tiempo estacionaria es aquella cuyas propiedades estadísticas, como la media y la varianza, son constantes a lo largo del tiempo. La integración se utiliza para transformar una serie de tiempo no estacionaria en una estacionaria aplicando diferenciación.

# En esta sesión trabajaremos con un modelo ARIMA

El modelo ARIMA (Autoregressive Integrated Moving Average) permite describir un valor como una función lineal de datos anteriores
y errores debidos al azar, además, puede incluir un componente cíclico o estacional. Es decir, debe contener todos los elementos
necesarios para describir el fenómeno. Box y Jenkins recomiendan como mínimo 50 observaciones en la serie temporal para aplicar
este modelo.

La metodología de Box y Jenkins se resume en cuatro fases:

1. **Primera fase: Identificación del modelo ARIMA**
   - En esta fase, se busca identificar el posible modelo ARIMA que sigue la serie temporal.
   - Para lograrlo, se aplican transformaciones a los datos para convertir la serie observada en una serie estacionaria.
   - Se busca determinar un modelo ARMA para la serie estacionaria. Este modelo incluye los órdenes $p$ y $q$ de su estructura autorregresiva y de media móvil respectivamente.

2. **Segunda fase: Estimación de parámetros**
   - Una vez seleccionado provisionalmente un modelo para la serie estacionaria, se pasa a la segunda etapa de estimación.
   - En esta etapa, los parámetros AR y MA del modelo se estiman utilizando el método de máxima verosimilitud. También se obtienen los errores estándar y los residuos del modelo.

3. **Tercera fase: Diagnóstico**
   - En esta fase se realiza un diagnóstico para verificar la adecuación del modelo.
   - Se comprueba que los residuos no tienen estructura de dependencia y siguen un proceso de ruido blanco. Si los residuos muestran alguna estructura, se modifica el modelo para incorporarla y se repiten las etapas anteriores hasta obtener un modelo adecuado.

4. **Cuarta fase: Predicción**
   - Una vez que se ha obtenido un modelo adecuado, se procede a realizar predicciones utilizando el mismo.






La metodología ARIMA se puede expresar matemáticamente de la siguiente manera:

Un modelo ARIMA($p$, $d$, $q$) se define por las siguientes ecuaciones:

- **Componente Autorregresivo (AR):**
  $$X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + \cdots + \phi_p X_{t-p} + \epsilon_t$$

- **Componente de Media Móvil (MA):**
  $$X_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q}$$

- **Componente Integrada (I):**
  $$X_t' = X_t - X_{t-1}$$

Donde:
- $X_t$ es el valor de la serie temporal en el tiempo $t$.
- $c$ es una constante.
- $\mu$ es la media de la serie.
- $\phi$ son los coeficientes del componente autorregresivo.
- $\theta$ son los coeficientes del componente de media móvil.
- $d$ representa el número de diferencias necesarias para hacer que la serie sea estacionaria.
- $\epsilon_t$ es el error en el tiempo $t$.

> Los modelos ARIMA solo se pueden aplicar en datos estacionarios. Eso significa que no queremos tener una tendencia en el tiempo. Si la serie de tiempo tiene una tendencia, entonces no es estacionaria y debemos aplicar la diferenciación para transformarla en estacionaria.

> También podemos usar la prueba de Dickey-Fuller para ayudarnos a concluir si la serie es estacionaria o no. La hipótesis nula de la prueba es que existe raíz unitaria con la alternativa de que no existe raíz unitaria.

## Prueba de Dickey-Fuller

La prueba de Dickey-Fuller es una herramienta estadística comúnmente utilizada para determinar si una serie temporal es estacionaria. Su formulación matemática es la siguiente:

> La hipótesis nula $H_{0}$ es que la serie temporal tiene una raíz unitaria, es decir, no es estacionaria. La hipótesis alternativa $H_{1}$ es que la serie temporal no tiene raíz unitaria, es decir, es estacionaria.

> $$
\text{H}_0: \text{No hay estacionariedad} \;
vs \;
\text{H}_1: \text{Hay estacionariedad}
$$

Para rechazar la hipótesis nula y concluir que la serie es estacionaria, necesitamos un valor $p$ ($p-value$) menor que un nivel de significancia dado, por ejemplo, $0.05$.


# Interpretación de la prueba Dickey-Fuller

**La prueba de Dickey-Fuller nos dice que cuando existe tendencia en una serie temporal en un modelo AR(1), el primer regresor tenderá a ser 1 o muy próximo a 1. Esto se debe a la propiedad de reversión a la media de un proceso estocástico estacionario.**

**En otras palabras, cuanto más cerca esté el primer coeficiente de un modelo AR(1) de 1, más tardarán las observaciones a volver al valor medio. Esto es sinónimo de no estacionariedad ya que, si el proceso estocástico fuera estable, este coeficiente sería menor a 1 o muy próximo a 0.**


```python
# Realizar la prueba de Dickey-Fuller para verificar la estacionalidad de la serie
result = adfuller(df['#Passengers'])
# Convertir los resultados de la prueba en un diccionario para facilitar su lectura
result_dict = dict(zip(['Test (ADF Test)', 'pvalue', 'usedlag', 'nobs',
                        'critical' 'values', 'icbest'], result))

# Imprimir los resultados de la prueba de Dickey-Fuller
print(result_dict)
```

| La función de autocorrelación (ACF) es una medida estadística utilizada en el análisis de series temporales. |
| En otras palabras, la ACF mide la correlación entre una serie temporal y una versión rezagada de sí misma en diferentes intervalos de tiempo. |
| La ACF se calcula para diferentes rezagos (o retrasos) y se traza en un gráfico llamado función de autocorrelación. |
| Este gráfico muestra la correlación entre la serie temporal en el tiempo actual y su valor en un tiempo anterior, desplazado por un número específico de períodos. |
| La ACF es útil para identificar patrones estacionales, cíclicos o de otro tipo en los datos, lo que puede ayudar en el proceso de modelado y predicción de series temporales. |

La fórmula matemática para la función de autocorrelación (ACF) se define como:

$$
\rho_k = \frac{\text{Cov}(y_t, y_{t-k})}{\sqrt{\text{Var}(y_t) \cdot \text{Var}(y_{t-k})}}
$$

donde:
- $ \rho_k $ es el coeficiente de autocorrelación en el rezago $ k $.
- $ \text{Cov}(y_t, y_{t-k}) $ es la covarianza entre la serie en el tiempo actual y su valor en el tiempo $ t-k $.
- $ \text{Var}(y_t) $ y $ \text{Var}(y_{t-k}) $ son las varianzas de la serie en los tiempos $ t $ y $ t-k $, respectivamente.

# Entrenamiento del modelo

```python
# Convertir los valores del DataFrame en un array
X = df.values

# Dividir los datos en conjunto de entrenamiento (75%) y de prueba (25%)
size = int(len(X) * 0.75)
train, test = X[0:size], X[size:len(X)]

# Crear una lista para almacenar el historial de datos de entrenamiento
history = [x for x in train]

# Lista para almacenar las predicciones
predictions = list()

```

```python
# Realizar la predicción paso a paso en el conjunto de prueba
for t in range(len(test)):
    # Ajustar el modelo ARIMA al historial de datos
    model_train = ARIMA(history, order=(2,1,1))
    modelo_train_fit = model_train.fit()

    # Realizar la predicción para el siguiente paso
    output = modelo_train_fit.forecast(steps=1)
    yhat = output[0]

    # Almacenar la predicción
    predictions.append(yhat)

    # Obtener la observación real del conjunto de prueba
    obs = test[t]

    # Añadir la observación real al historial para la próxima iteración
    history.append(obs)

```

```python
# Convertir las predicciones en un array y redimensionarlo
fc = np.array(predictions).reshape(-1, 1)

# Combinar el conjunto de entrenamiento y prueba completo
test_full = np.concatenate([train, test])

# Combinar el conjunto de entrenamiento y las predicciones
predictions_full = np.concatenate([train, fc])

# Mostrar los resultados en una gráfica
plt.plot(test_full, color='green', label='Datos Reales')
plt.plot(predictions_full, color='red', label='Predicciones')
plt.legend()
plt.title('Pronóstico del Modelo ARIMA')
plt.xlabel('Tiempo')
plt.ylabel('Número de Pasajeros')
plt.show()
```

### Explicación del AIC

**AIC (Akaike Information Criterion)**:

El AIC es una medida utilizada para seleccionar el mejor modelo entre un conjunto de modelos candidatos. Se basa en la teoría de la información y busca el modelo que mejor explica los datos con el menor número de parámetros.

El AIC se calcula con la fórmula:

$$
AIC = 2k - 2\ln(L)
$$

donde:

- $k$ es el número de parámetros en el modelo.
- $L$ es la máxima verosimilitud del modelo.

En la selección de modelos, un valor más bajo de AIC indica un modelo mejor. Sin embargo, es importante no interpretar el AIC en términos absolutos, sino compararlo entre modelos ajustados con los mismos datos.

### Explicación de la Tabla

- **Modelo**: La especificación del modelo ARIMA utilizado. El formato es ARIMA(p,d,q)(P,D,Q)[m], donde:
  - \(p\): Orden del término autorregresivo (AR).
  - \(d\): Número de diferenciaciones necesarias para hacer la serie estacionaria.
  - \(q\): Orden del término de media móvil (MA).
  - \(P\): Orden del componente autorregresivo estacional.
  - \(D\): Número de diferenciaciones estacionales.
  - \(Q\): Orden del componente de media móvil estacional.
  - \([m]\): Período estacional.

- **AIC**: Akaike Information Criterion, un valor más bajo de AIC indica un modelo mejor. Es una medida utilizada para seleccionar el mejor modelo entre un conjunto de modelos candidatos. Se basa en la teoría de la información y busca el modelo que mejor explica los datos con el menor número de parámetros. La fórmula es:

  $$ \text{AIC} = 2k - 2\ln(L) $$

  donde:
  - \(k\) es el número de parámetros en el modelo.
  - \(L\) es la máxima verosimilitud del modelo.

- **Tiempo**: El tiempo que tomó ajustar el modelo en segundos.
