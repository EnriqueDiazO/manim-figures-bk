## 7.1. Módulos os y sys

- **os:** Proporciona funciones para interactuar con el sistema operativo, como manipulación de archivos, rutas, etc.
- **sys:** Proporciona funciones y variables que interactúan fuertemente con el intérprete de Python.



### Módulo os:

El módulo os en Python proporciona una interfaz para interactuar con el sistema operativo subyacente. Algunas de las funcionalidades más comunes que ofrece incluyen:

    - Manipulación de archivos y directorios.
    - Obtención de información sobre el sistema operativo.
    - Manipulación de rutas de archivos.



## Manipulación de archivos y directorios.

````python
import os
# Obtener el directorio de trabajo actual
directorio_actual = os.getcwd()
print("Directorio actual:", directorio_actual)
````

````python
try:
    # Crear un directorio
    os.mkdir("nuevo_directorio")
    print("Se ha creado el directorio")
except FileExistsError:
    print("El directorio ya existe")
````

````python
nuevo_directorio = "nuevo_directorio"
try:
    # Cambiar al nuevo directorio
    os.chdir(nuevo_directorio)
    print("Se ha cambiado al siguiente directorio:", os.getcwd())
except FileNotFoundError:
    print("El directorio especificado no existe")
````

````python
nombre_archivo = "texto.txt"
contenido = "Este es el contenido del archivo de texto."
try:
    ruta_archivo = os.path.join(os.getcwd(), nombre_archivo)

    # Crear y escribir en el archivo
    with open(ruta_archivo, "w") as archivo:
        archivo.write(contenido)

    print(f"Se ha creado el archivo '{nombre_archivo}' con el siguiente contenido:")
    print(contenido)

except Exception as e:
    print(f"No se pudo crear el archivo '{nombre_archivo}': {e}")
````

````python
try:
    # Cambiar el nombre de un archivo
    os.rename("texto.txt", "nuevo_texto.txt")
    print("Se ha cambiado el nombre del archivo correctamente.")
except Exception as e:
    print(f"No se pudo cambiar el nombre del archivo: {e}")
````

````python
try:
    # Eliminar un archivo
    os.remove("nuevo_texto.txt")
    print("El archivo 'nuevo_texto.txt' ha sido eliminado correctamente.")
except Exception as e:
    print(f"No se pudo eliminar el archivo 'nuevo_texto.txt': {e}")
````

````python
try:
    # Obtener el directorio actual
    directorio_actual = os.getcwd()
    print(f"Directorio actual: {directorio_actual}")

    # Obtener la ruta del directorio padre
    directorio_padre = os.path.dirname(directorio_actual)
    print(f"Directorio padre: {directorio_padre}")

    # Cambiar al directorio padre
    os.chdir(directorio_padre)

    # Imprimir el nuevo directorio actual
    print("Se ha cambiado al directorio padre:", os.getcwd())

except Exception as e:
    print(f"No se pudo cambiar al directorio padre: {e}")
````

````python
# Directorio que deseas eliminar
directorio_a_borrar = "nuevo_directorio"

# Verificar si el directorio existe antes de eliminarlo
if os.path.exists(directorio_a_borrar):
    # Eliminar el directorio
    os.rmdir(directorio_a_borrar)
    print(f"El directorio '{directorio_a_borrar}' ha sido eliminado correctamente.")
else:
    print(f"El directorio '{directorio_a_borrar}' no existe.")
````

## Manipulación de rutas de archivos.

````python
import os

# Unir rutas
ruta_completa = os.path.join("directorio", "subdirectorio", "archivo.txt")
print("Ruta completa:", ruta_completa)
````

````python
# Dividir una ruta en componentes
ruta = "/directorio/subdirectorio/archivo.txt"
componentes = os.path.split(ruta)
print("Componentes de la ruta:", componentes)
# Observa que los resultados los devuelve en forma de tupla,
````

````python
# Obtener el nombre del archivo de una ruta
nombre_archivo = os.path.basename("/directorio/subdirectorio/archivo.txt")
print("Nombre del archivo:", nombre_archivo)
````

````python
# Obtener el directorio padre de una ruta
directorio_padre = os.path.dirname("/directorio/subdirectorio/archivo.txt")
print("Directorio padre:", directorio_padre)
````

## 7.2. Objetos File

Los objetos de archivo en Python proporcionan una forma conveniente de trabajar con archivos en disco.

### Objetos de archivo:

En Python, los objetos de archivo representan archivos en el sistema de archivos del sistema operativo subyacente. Proporcionan una interfaz para interactuar con los archivos, permitiendo operaciones como lectura, escritura, cierre y manipulación de la posición del puntero dentro del archivo.

````python
# Ejemplo de creación de un objeto de archivo
with open("archivo.txt", "w") as archivo:
    archivo.write("Contenido de ejemplo")
````

````python
# Ejemplo de lectura desde un archivo
with open("archivo.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)
````

| Modo  | Descripción                                           | Ejemplo                                       |
|-------|-------------------------------------------------------|-----------------------------------------------|
| "r"   | Lectura. Abre el archivo para lectura.               | `archivo = open("archivo.txt", "r")`          |
| "w"   | Escritura. Abre el archivo y trunca su contenido.   | `archivo = open("archivo.txt", "w")`          |
| "a"   | Adición. Abre el archivo para añadir contenido.      | `archivo = open("archivo.txt", "a")`          |
| "r+"  | Lectura y escritura. Abre el archivo para leer y escribir, comienza al principio. | `archivo = open("archivo.txt", "r+")` |
| "w+"  | Lectura y escritura. Abre el archivo para leer y escribir, trunca el contenido. | `archivo = open("archivo.txt", "w+")` |
| "a+"  | Lectura y adición. Abre el archivo para leer y añadir contenido. | `archivo = open("archivo.txt", "a+")` |


## 7.2.1. Apertura

En Python, para trabajar con archivos, primero debes abrirlos. Puedes abrir un archivo en diferentes modos, como lectura, escritura o ambos. La función `open()` se utiliza para abrir un archivo y devuelve un objeto de archivo que puedes usar para realizar operaciones en el archivo.

````python
### Ejemplo de código:
# Abrir un archivo en modo lectura
with open("archivo.txt", "r") as archivo:
    # Realizar operaciones de lectura aquí
    pass
````

## 7.2.2. Lectura

Una vez que has abierto un archivo en modo lectura, puedes leer su contenido utilizando métodos como read(), readline() o readlines(). Estos métodos te permiten leer el contenido del archivo completo, una línea a la vez o todas las líneas a la vez, respectivamente.

````python
# Leer todo el contenido del archivo
with open("archivo.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)
````

## 7.2.3. Escritura

Cuando abres un archivo en modo escritura, puedes escribir datos en él utilizando el método write(). Puedes escribir cadenas de texto u otros tipos de datos en el archivo. Si el archivo no existe, se creará. Si ya existe, su contenido se sobrescribirá.

````python
# Escribir en un archivo
with open("nuevo_archivo.txt", "w") as archivo:
    archivo.write("Este es un nuevo archivo de texto.")
````

## 7.2.4. Posición

El objeto de archivo en Python tiene un puntero interno que indica la posición actual dentro del archivo. Puedes usar el método tell() para obtener la posición actual del puntero de lectura/escritura dentro del archivo. También puedes usar el método seek() para cambiar la posición del puntero.


````python
# Obtener la posición actual dentro del archivo
with open("archivo.txt", "r") as archivo:
    posicion = archivo.tell()
    print("Posición actual:", posicion)
````
