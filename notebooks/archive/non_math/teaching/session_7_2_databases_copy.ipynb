# Tema 8: Introducción a bases de datos
## Objetivo general:
**Objetivo General**:

- Proporcionar a los estudiantes una comprensión fundamental de las bases de datos, incluyendo sus componentes básicos como tablas, campos y tipos de datos, así como familiarizarlos con el lenguaje SQL y su sintaxis, permitiéndoles realizar operaciones básicas como la creación, modificación y consulta de datos en una base de datos relacional.

**Objetivos Específicos**:
- Comprender los conceptos fundamentales de las bases de datos, incluyendo la estructura de tablas y campos.

- Reconocer los diferentes tipos de datos utilizados en las bases de datos relacionales y su aplicación.

- Familiarizarse con el lenguaje SQL y su sintaxis básica para realizar consultas simples.

- Desarrollar habilidades para realizar consultas básicas en una base de datos utilizando sentencias SQL como SELECT, INSERT, UPDATE y DELETE.

### Glosario

- **Bases de datos**: Una base de datos es una recopilación organizada de información o datos estructurados, que
normalmente se almacena de forma electrónica en un sistema informático.

- **Tablas**: Los datos de los tipos más comunes de bases de datos en funcionamiento actualmente se suelen utilizar como estructuras de filas y columnas en una serie de tablas para aumentar la eficacia del procesamiento y la consulta de datos.

- **Campo**: En el contexto de las bases de datos, un campo se refiere a una unidad básica de datos que representa una característica específica de una entidad en una tabla. Cada fila en una tabla está compuesta por uno o más campos, y cada campo tiene un nombre y un tipo de datos asociado que define qué tipo de información puede almacenar.

- **Tipo de datos**: En SQL, los tipos de datos pueden variar según el sistema de gestión de bases de datos (DBMS) que estés utilizando. Sin embargo, algunos tipos de datos comunes incluyen INTEGER, VARCHAR, DATE, FLOAT, entre otros.

| Tipo de dato  | Descripción                                                   | Ejemplo                           |
|---------------|---------------------------------------------------------------|-----------------------------------|
| INTEGER       | Números enteros                                               | `id INTEGER PRIMARY KEY`          |
| VARCHAR(n)    | Cadenas de caracteres de longitud variable, hasta n caracteres | `name VARCHAR(50)`                |
| CHAR(n)       | Cadenas de caracteres de longitud fija, exactamente n caracteres | `initial CHAR(1)`                 |
| DATE          | Fechas                                                        | `birth_date DATE`                 |
| TIME          | Tiempos                                                       | `check_in TIME`                   |
| TIMESTAMP     | Fecha y tiempo                                                | `last_updated TIMESTAMP`          |
| BOOLEAN       | Valores booleanos (TRUE o FALSE)                              | `is_active BOOLEAN`               |
| FLOAT         | Números decimales                                             | `salary FLOAT`                    |
| DECIMAL(p,s)  | Números decimales con precisión fija                          | `price DECIMAL(10,2)`             |
| ARRAY         | Arreglos o matrices de datos                                 | `items ARRAY`                     |



- **Sistema de bases de datos**: Un sistema de base de datos suele contener varias bases de datos. Cada base de datos suele
contener varias tablas. Las tablas almacenan los datos organizados por filas. Cada fila contiene varios campos. Cada campo tiene una serie de propiedades, como por ejemplo, el nombre del campo, su longitud, tipo de dato que se va a almacenar, etc.


- **SQL(Structured Query Language, lenguaje de consulta estructurada)**: El lenguaje SQL es un lenguaje estándar utilizado para interactuar con bases de datos relacionales. Permite realizar diversas operaciones, como la creación, modificación y consulta de datos en una base de datos.

- **Sintaxis de SQL**: Las sentencias SQL no son sensibles a mayúsculas y minúsculas, es decir, 'SELECT' y 'select'
son dos palabras iguales para SQL.

## 8.1. El lenguaje SQL

### Características principales:

- **Lenguaje Declarativo**: SQL es un lenguaje declarativo, lo que significa que describes el resultado deseado y no el proceso para obtenerlo. Esto permite a los usuarios centrarse en qué datos necesitan en lugar de cómo obtenerlos.

- **Basado en Conjuntos**: SQL está diseñado para trabajar con conjuntos de datos en lugar de operar en registros individuales. Esto permite realizar operaciones eficientes en grandes volúmenes de datos.

- **Independiente de la Plataforma**: SQL es compatible con una amplia variedad de sistemas de gestión de bases de datos (DBMS), lo que lo hace altamente portátil y permite la interoperabilidad entre diferentes sistemas.


| Componente                        | Descripción                                                                                                      | Ejemplo de SQL                                                                          |
|----------------------------------|------------------------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------|
| DDL (Data Definition Language)   | Se utiliza para definir y modificar la estructura de la base de datos, incluyendo la creación, modificación y eliminación de bases de datos, tablas, índices, etc. | `CREATE TABLE employees (id INT PRIMARY KEY, name VARCHAR(50), age INT);`                |
| DML (Data Manipulation Language) | Se utiliza para manipular los datos en la base de datos, incluyendo la inserción, actualización y eliminación de registros en tablas. | `INSERT INTO employees (id, name, age) VALUES (1, 'John', 30);`                           |
| DQL (Data Query Language)        | Se utiliza para realizar consultas para recuperar información de la base de datos. La consulta más común es la sentencia SELECT. | `SELECT * FROM employees WHERE age > 25;`                                                |
| DCL (Data Control Language)      | Se utiliza para administrar los permisos de acceso a los datos en la base de datos, incluyendo la concesión y revocación de privilegios. | `GRANT SELECT ON employees TO user1;`                                                    |

## 8.2. Sintaxis básica de las sentencias SQL

Las sentencias SQL siguen una estructura básica que incluye palabras clave como SELECT, INSERT, UPDATE, DELETE, CREATE, DROP, entre otras. Estas sentencias se utilizan para realizar operaciones específicas en la base de datos.

| Sentencia SQL | Descripción                                                                                                            | Ejemplo                                                                        |
|---------------|------------------------------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------|
| SELECT        | Se utiliza para recuperar datos de una o más tablas de la base de datos.                                              | `SELECT column1, column2, ... FROM table_name WHERE condition;`                |
| INSERT        | Se utiliza para agregar nuevos registros a una tabla.                                                                  | `INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);` |
| UPDATE        | Se utiliza para modificar registros existentes en una tabla.                                                           | `UPDATE table_name SET column1 = value1, column2 = value2, ... WHERE condition;` |
| DELETE        | Se utiliza para eliminar registros de una tabla.                                                                       | `DELETE FROM table_name WHERE condition;`                                      |
| CREATE        | Se utiliza para crear objetos en la base de datos, como tablas, índices, vistas, etc.                                  | `CREATE TABLE table_name (column1 datatype, column2 datatype, ...);`            |
| DROP          | Se utiliza para eliminar objetos de la base de datos.                                                                 | `DROP TABLE table_name;`                                                      |

Por ejemplo:
```sql
CREATE TABLE usuarios (
    id INTEGER PRIMARY KEY,
    nombre TEXT,
    edad INTEGER
);
```

## CREATE TABLE AS
La sintaxis en SQL para copiar una tabla utilizando la instrucción CREATE TABLE AS es la siguiente:
```sql
CREATE TABLE nueva_tabla AS
SELECT *
FROM tabla_original;
```
Por ejemplo:

```sql
CREATE TABLE usuarios_nueva AS
SELECT *
FROM usuarios;
```

## Añadir registros

Para agregar información a una tabla en SQL, utilizamos la instrucción INSERT INTO. Aquí tienes la sintaxis básica:

```sql
INSERT INTO nombre_de_la_tabla (columna1, columna2, ...)
VALUES (valor_columna1, valor_columna2, ...);
```
Por ejemplo:

```sql
INSERT INTO usuarios (nombre, edad) VALUES
('Juan', 25),
('María', 30),
('Pedro', 40),
('Ana', 35),
('Luis', 28);
```

## ALTER TABLE

La sentencia ALTER TABLE se utiliza para modificar la estructura de una tabla existente, como agregar o eliminar columnas. Su estructura básica es:

```sql
ALTER TABLE table_name
ADD column_name datatype;

ALTER TABLE table_name
DROP COLUMN column_name;


```


Por ejemplo
```sql
ALTER TABLE usuarios
ADD email TEXT;
```

```sql
ALTER TABLE usuarios
DROP COLUMN edad;
```




## DROP TABLE

La sentencia DROP TABLE se utiliza para eliminar una tabla existente de la base de datos. Su estructura básica es:

```sql
DROP TABLE table_name;
```

Por ejemplo:
```sql
DROP TABLE usuarios;
```

## 8.4. Insertar datos en una tabla

Para agregar datos a una tabla existente, se utiliza la sentencia SQL INSERT INTO, que permite especificar los valores que se van a insertar en cada columna de la tabla.
INSERT INTO

La sentencia INSERT INTO se utiliza para agregar nuevos registros a una tabla existente en la base de datos. Su estructura básica es la siguiente:


```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```



Por ejemplo, si tenemos una tabla llamada empleados con columnas id, nombre, y salario, podríamos insertar un nuevo empleado de la siguiente manera:
```sql
INSERT INTO empleados (id, nombre, salario)
VALUES (1, 'Juan', 30000);
```

Esta sentencia insertaría un nuevo registro en la tabla empleados con el id igual a 1, el nombre igual a 'Juan', y el salario igual a 30000.

## 8.5. Seleccionar registros

La operación más común en SQL es la selección de datos de una tabla. Esto se logra utilizando la sentencia SELECT, que permite recuperar uno o más registros que cumplen con ciertos criterios.


### SELECT

La sentencia SELECT se utiliza para recuperar datos de una o más tablas de la base de datos. Su estructura básica es la siguiente:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition;
```

Por ejemplo, si tenemos una tabla llamada empleados con columnas id, nombre, y salario, podríamos seleccionar todos los empleados cuyo salario sea mayor a 50000 de la siguiente manera:

```sql
-- 1. Agregar la columna 'Salary' a la tabla 'Employee'
ALTER TABLE Employee
ADD COLUMN Salary FLOAT;

-- 2. Actualizar la nueva columna 'Salary' con valores aleatorios en un rango  (0 ,9999)
UPDATE Employee
SET Salary = ABS(RANDOM()) % 100000;

-- 3. Seleccionar empleados
SELECT *
FROM Employee
WHERE Salary > 5000;
```

Esta sentencia seleccionaría los id y nombre de todos los empleados cuyo salario sea mayor a 50000.

## 8.6. Modificar registros

Para actualizar datos existentes en una tabla, se utiliza la sentencia UPDATE. Esta sentencia permite modificar los valores de una o varias columnas en función de ciertas condiciones.


### UPDATE

La sentencia UPDATE se utiliza para modificar registros existentes en una tabla. Su estructura básica es la siguiente:

```sql
UPDATE table_name
SET column1 = value1, column2 = value2, ...
WHERE condition;

```

Por ejemplo, si queremos aumentar el salario de un empleado con el id igual a 1 en una tabla llamada empleados, podríamos hacerlo de la siguiente manera:

```sql
UPDATE Employee
SET Salary = Salary * 1.1
WHERE id = 1;
```

## 8.7. Eliminar registros

Para eliminar datos de una tabla, se utiliza la sentencia DELETE FROM, que permite eliminar uno o más registros que cumplen con ciertos criterios.

### DELETE FROM

La sentencia DELETE FROM se utiliza para eliminar registros de una tabla. Su estructura básica es la siguiente:

```sql
DELETE FROM table_name
WHERE condition;
```
Por ejemplo, si queremos eliminar un empleado con el id igual a 1 de una tabla llamada empleados, podríamos hacerlo de la siguiente manera:

```sql
DELETE FROM Employee
WHERE id = 2;
```
Esta sentencia eliminaría el empleado con id igual a 1 de la tabla empleados.


### Es importante tener cuidado al utilizar la sentencia DELETE FROM, ya que si se ejecuta sin una condición adecuada, puede eliminar todos los registros de la tabla, lo que resultaría en la pérdida completa de datos

#### Para eliminar la columna empleados
```sql
DELETE FROM empleados;
```
#### Para eliminar toda la tabla
```sql
DELETE FROM table_name;
```

## 8.8. Consultas

Las consultas en SQL permiten combinar diferentes operaciones para recuperar información específica de una base de datos. Esto se logra utilizando cláusulas como WHERE, GROUP BY, ORDER BY, JOIN, entre otras.


#### SELECT con cláusula WHERE

- La cláusula WHERE se utiliza para filtrar los resultados de una consulta basada en ciertos criterios. Por ejemplo:

```sql
SELECT column1, column2
FROM table_name
WHERE condition;
```
Esto seleccionaría las columnas column1 y column2 de la tabla table_name donde se cumpla la condición especificada.


Por ejemplo:
```sql
SELECT *
FROM Employee
WHERE strftime('%Y', Birthdate) = '1962';
```




#### GROUP BY

- La cláusula GROUP BY se utiliza para agrupar filas que tienen el mismo valor en una o más columnas. Por ejemplo:

```sql
SELECT column1, COUNT(*)
FROM table_name
GROUP BY column1;
```
Por ejemplo

```sql
SELECT City, COUNT(*)
FROM Employee
GROUP BY City;
```


#### ORDER BY
- La cláusula ORDER BY se utiliza para ordenar los resultados de una consulta en función de una o más columnas. Por ejemplo:

```sql
SELECT column1, column2
FROM table_name
ORDER BY column1 ASC, column2 DESC;
```

Por ejemplo:

```sql
SELECT LastName, FirstName, Salary
FROM Employee
ORDER BY LastName ASC, Salary DESC;
```
#### JOIN

La cláusula JOIN se utiliza para combinar filas de dos o más tablas en base a una condición relacionada entre ellas. Por ejemplo:
```sql
SELECT column1, column2
FROM table1
JOIN table2 ON table1.id = table2.id;
```

Esto combinaría las filas de las tablas table1 y table2 donde los valores de la columna id sean iguales en ambas tablas.

```sql
-- Selecciona los apellidos del empleado, los nombres del empleado,
-- los apellidos del cliente, los nombres del cliente y la ciudad del cliente.
SELECT e.LastName AS EmployeeLastName, e.FirstName AS EmployeeFirstName
       c.LastName AS CustomerLastName, c.FirstName AS CustomerFirstName, c.City
       
-- De la tabla Employee (empleados)
FROM Employee e

-- Une la tabla Employee (empleados) con la tabla Customer (clientes) utilizando la columna LastName
JOIN Customer c ON e.LastName = c.LastName;

```
