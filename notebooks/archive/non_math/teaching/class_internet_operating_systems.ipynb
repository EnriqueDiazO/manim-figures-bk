# **Protocolo:** Conjunto de reglas convenido.

>##  **TCP/IP (Transfer Control Protocol/Internet Protocol):** IP es la parte que obtiene la dirección a la que se envían los datos. TCP se encarga de la entrega de los datos una vez hallada dicha dirección IP. 
>## Ejemplo: La dirección IP es como el número de teléfono que se asigna a su smartphone. TCP es toda la tecnología que hace que el teléfono emita un timbre al recibir una llamada y que le permite hablar con alguien al otro lado de la línea.

>## **TCP/IP divide las distintas tareas de comunicación en capas**. Cada capa tiene una función distinta. Los datos pasan por cuatro capas independientes antes de recibirse en el otro extremo (como se explica en la sección siguiente). A continuación, TCP/IP recorre estas capas en orden inverso para reensamblar los datos y presentárselos al destinatario.

> # **Sistema Operativo**: Sistema complejo usado para administrar archivos, dispositivos de entrada y salida, procesos, memoria, etc. Cada elemento de este sistema está bien definido y clasificado para realizar sus operaciones

> ![picture](https://drive.google.com/uc?id=1UW8BAIUALY88fmz4wPYZw8TTKfP3gWlc)
[Link](https://drive.google.com/uc?id=1UW8BAIUALY88fmz4wPYZw8TTKfP3gWlc)

> # **Kernel**: Se encarga del nivel básico de control de todos los periféricos de la computadora, esto es, se encarga de conceder el acceso al hardware de forma segura para todo el software que lo solicita.

> # **API**: API significa interfaz de programación de aplicaciones, esto es un conjunto de definiciones y protocolos que se utiliza para desarrollar e integrar el software de las aplicaciones

> # **Sistema de Archivos**: Se encarga de cómo y dónde se almacenan los datos en el disco, ya sea físico o lógico. 


> # **Controladores de dispositivos**: Se encargan de ser intermediarios entre el sistema operativo y el periférico que se desea implementar.


> # **Dispositivos de hardware**: Son todos los elementos tangibles de nuestra computadora que serán usados por el sistema operativo. Entre estos están los monitores, bocinas, mouse, teclado, impresoras, discos de almacenamiento, memorias externas etc.
