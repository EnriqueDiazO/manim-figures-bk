```python
from gtts import gTTS
from pydub import AudioSegment

# Generar audios TTS individuales
gTTS("mamá", lang='es').save("mamá.mp3")
gTTS("papá", lang='es').save("papá.mp3")
gTTS("hola", lang='es').save("hola.mp3")

# Cargar los audios generados
mamá_seg = AudioSegment.from_mp3("mamá.mp3").set_frame_rate(16000)
papá_seg = AudioSegment.from_mp3("papá.mp3").set_frame_rate(16000)
hola_seg = AudioSegment.from_mp3("hola.mp3").set_frame_rate(16000)

# Recortar o alargar artificialmente para que duren aprox. 700 ms cada uno
mamá_seg = mamá_seg[:700]
papá_seg = papá_seg[:700]
hola_seg = hola_seg[:700]

# Silencios entre palabras para que encajen con los tiempos deseados
sil_0_0_5 = AudioSegment.silent(duration=500)   # 0.0s – 0.5s
sil_1_2_1_5 = AudioSegment.silent(duration=300) # 1.2s – 1.5s
sil_2_2_2_5 = AudioSegment.silent(duration=300) # 2.2s – 2.5s

# Unir todos los segmentos
audio_final = (
    sil_0_0_5 + mamá_seg +
    sil_1_2_1_5 + papá_seg +
    sil_2_2_2_5 + hola_seg
)

# Exportar como audio_001.wav
audio_final.export("audio_001.wav", format="wav")
```
