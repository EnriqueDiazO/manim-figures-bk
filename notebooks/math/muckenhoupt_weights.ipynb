```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Configuración general
# ------------------------------------------------------------
plt.rcParams.update({
    "font.size": 16,
    "mathtext.fontset": "cm",
    "font.family": "serif",
    "axes.spines.top": False,
    "axes.spines.right": False,
})

# Dominio
x = np.linspace(-2.2, 2.2, 2000)

# ------------------------------------------------------------
# Utilidad para dar formato homogéneo
# ------------------------------------------------------------
def setup_axes(ax, ylabel):
    ax.spines["left"].set_position("zero")
    ax.spines["bottom"].set_position("zero")
    ax.set_xlim(-2.2, 2.2)
    ax.set_xticks([-2, -1, 0, 1, 2])
    ax.set_xlabel(r"$x$", loc="right")
    ax.set_ylabel(ylabel, rotation=0, labelpad=25)
    ax.yaxis.set_label_coords(0.08, 0.82)
    ax.tick_params(direction="in", length=4, width=1)

# ------------------------------------------------------------
# 1) delta > 0
# Usamos delta = 1/2 como representante visual
# ------------------------------------------------------------
delta_pos = 0.4
y_pos = np.abs(x) ** delta_pos

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y_pos, linewidth=2)
setup_axes(ax, r"$w_\delta(x)$")
ax.set_ylim(0, 1.6)
ax.set_yticks([0.5, 1.0, 1.5])
ax.set_title(r"$\delta>0$", pad=12)
fig.tight_layout()
fig.savefig("weight_delta_positive.png", dpi=300, bbox_inches="tight")
plt.close(fig)

# ------------------------------------------------------------
# 2) delta = 0
# ------------------------------------------------------------
y_zero = np.ones_like(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y_zero, linewidth=2)
setup_axes(ax, r"$w_0(x)$")
ax.set_ylim(0, 1.6)
ax.set_yticks([0.5, 1.0, 1.5])
ax.set_title(r"$\delta=0$", pad=12)
fig.tight_layout()
fig.savefig("weight_delta_zero.png", dpi=300, bbox_inches="tight")
plt.close(fig)

# ------------------------------------------------------------
# 3) delta < 0
# Usamos delta = -1/2 como representante visual
# Evitamos x=0 para que no explote en el dibujo
# ------------------------------------------------------------
delta_neg = -0.5
eps = 0.05
x_left = np.linspace(-2.2, -eps, 1000)
x_right = np.linspace(eps, 2.2, 1000)
y_left = np.abs(x_left) ** delta_neg
y_right = np.abs(x_right) ** delta_neg

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x_left, y_left, linewidth=2)
ax.plot(x_right, y_right, linewidth=2)
ax.axvline(0, linestyle="--", linewidth=1)
setup_axes(ax, r"$w_\delta(x)$")
ax.set_ylim(0, 4.5)
ax.set_yticks([1, 2, 3, 4])
ax.set_title(r"$\delta<0$", pad=12)
fig.tight_layout()
fig.savefig("weight_delta_negative.png", dpi=300, bbox_inches="tight")
plt.close(fig)

print("Archivos generados:")
print(" - weight_delta_positive.png")
print(" - weight_delta_zero.png")
print(" - weight_delta_negative.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Estilo general
# ------------------------------------------------------------
plt.rcParams.update({
    "font.size": 16,
    "font.family": "serif",
    "mathtext.fontset": "cm",
    "axes.spines.top": False,
    "axes.spines.right": False,
})

BLUE = "#1f4fa3"
GREEN = "#2f8f3a"
RED = "#d62728"

# ------------------------------------------------------------
# Utilidad para dar formato homogéneo
# ------------------------------------------------------------
def setup_axes(ax, ylabel, ylim, yticks, vline_zero=False, spine_color="black"):
    ax.spines["left"].set_position("zero")
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_color(spine_color)
    ax.spines["bottom"].set_color(spine_color)
    ax.spines["left"].set_linewidth(1.2)
    ax.spines["bottom"].set_linewidth(1.2)

    ax.set_xlim(-2.35, 2.35)
    ax.set_ylim(*ylim)

    ax.set_xticks([-2, -1, 0, 1, 2])
    ax.set_yticks(yticks)

    ax.tick_params(direction="in", length=4, width=1, colors="black")

    ax.set_xlabel(r"$x$", fontsize=22)
    ax.set_ylabel(ylabel, fontsize=22, rotation=0)

    ax.xaxis.set_label_coords(1.03, 0.05)
    ax.yaxis.set_label_coords(0.06, 0.82)

    if vline_zero:
        ax.axvline(0, color="gray", linestyle="--", linewidth=1.5, zorder=0)

# ------------------------------------------------------------
# 1) delta > 0
# ------------------------------------------------------------
x = np.linspace(-2.2, 2.2, 2000)
delta_pos = 2.0
y_pos = (np.abs(x) / 2.0) ** delta_pos

fig, ax = plt.subplots(figsize=(6.4, 4.2))
ax.plot(x, y_pos, color=BLUE, linewidth=2.2)

setup_axes(
    ax,
    ylabel=r"$w_\delta(x)$",
    ylim=(0, 1.45),
    yticks=[0.5, 1.0],
    spine_color=BLUE
)

ax.set_title(r"$\delta>0$ (vanishes at 0)", color=BLUE, fontsize=20, pad=14)
fig.tight_layout()
fig.savefig("weight_delta_positive.png", dpi=300, bbox_inches="tight")
plt.close(fig)

# ------------------------------------------------------------
# 2) delta = 0
# ------------------------------------------------------------
y_zero = np.ones_like(x)

fig, ax = plt.subplots(figsize=(6.4, 4.2))
ax.plot(x, y_zero, color=GREEN, linewidth=2.2)

setup_axes(
    ax,
    ylabel=r"$w_0(x)$",
    ylim=(0, 1.45),
    yticks=[0.5, 1.0],
    vline_zero=True
)

ax.set_title(r"$\delta=0$ (constant weight)", color=GREEN, fontsize=20, pad=14)
fig.tight_layout()
fig.savefig("weight_delta_zero.png", dpi=300, bbox_inches="tight")
plt.close(fig)

# ------------------------------------------------------------
# 3) delta < 0
# ------------------------------------------------------------
delta_neg = -0.5
eps = 0.05
x_left = np.linspace(-2.2, -eps, 1400)
x_right = np.linspace(eps, 2.2, 1400)
y_left = np.abs(x_left) ** delta_neg
y_right = np.abs(x_right) ** delta_neg

fig, ax = plt.subplots(figsize=(6.4, 4.2))
ax.plot(x_left, y_left, color=RED, linewidth=2.2)
ax.plot(x_right, y_right, color=RED, linewidth=2.2)

setup_axes(
    ax,
    ylabel=r"$w_\delta(x)$",
    ylim=(0, 1.45),
    yticks=[0.5, 1.0],
    vline_zero=True
)

ax.set_title(r"$\delta<0$ (blows up at 0)", color=RED, fontsize=20, pad=14)
fig.tight_layout()
fig.savefig("weight_delta_negative.png", dpi=300, bbox_inches="tight")
plt.close(fig)

print("Archivos generados:")
print("  weight_delta_positive.png")
print("  weight_delta_zero.png")
print("  weight_delta_negative.png")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Configuración global
# ============================================================

plt.rcParams.update({
    "figure.dpi": 160,
    "savefig.dpi": 300,
    "font.family": "serif",
    "mathtext.fontset": "cm",
    "axes.linewidth": 0.8,
})

OUTDIR = "figures"
os.makedirs(OUTDIR, exist_ok=True)


# ============================================================
# Utilidades geométricas
# ============================================================

def rotate(x, y, angle):
    """Rota el conjunto de puntos (x,y) un ángulo angle."""
    c, s = np.cos(angle), np.sin(angle)
    xr = c * x - s * y
    yr = s * x + c * y
    return xr, yr


def setup_ax(ax, xlim=(-3.2, 3.2), ylim=(-2.4, 2.4)):
    """Configura un eje limpio, sin ejes ni marcas."""
    ax.set_aspect("equal")
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.axis("off")


def endpoint(ax, x, y, size=48):
    """Marca un punto extremo."""
    ax.scatter([x], [y], s=size, color="black", zorder=5)


def common_point(ax, x=0.0, y=0.0, size=32):
    """Marca el punto común de la estrella."""
    ax.scatter([x], [y], s=size, color="black", zorder=6)


def save_fig(fig, filename):
    path = os.path.join(OUTDIR, filename)
    fig.savefig(path, bbox_inches="tight", pad_inches=0.03)
    plt.close(fig)
    print(f"Saved: {path}")


# ============================================================
# 1) Piecewise Lyapunov curve without cusps
# ============================================================

def draw_lyapunov_no_cusps():
    fig, ax = plt.subplots(figsize=(4.8, 3.0))
    setup_ax(ax, xlim=(-3.2, 3.2), ylim=(-1.7, 1.9))

    pts = np.array([
        [-2.8, -0.6],
        [-1.6, -0.25],
        [-0.7,  0.55],
        [ 0.2, -0.10],
        [ 1.3,  0.80],
        [ 2.6,  0.15],
    ])

    ax.plot(pts[:, 0], pts[:, 1], color="black", lw=3.0, solid_capstyle="round")
    endpoint(ax, pts[0, 0], pts[0, 1])
    endpoint(ax, pts[-1, 0], pts[-1, 1])

    save_fig(fig, "geometry_lyapunov_no_cusps.png")


# ============================================================
# 2) Piecewise Lyapunov curve with first-order cusp
#    Usamos una semicúbica: x=t^2, y=t^3
# ============================================================

def draw_lyapunov_first_order_cusps():
    fig, ax = plt.subplots(figsize=(4.8, 3.0))
    setup_ax(ax, xlim=(-3.2, 3.2), ylim=(-2.0, 2.0))

    t = np.linspace(-1.7, 1.7, 600)
    x = 1.45 * t**2 - 1.4
    y = 0.80 * t**3

    # ligera rotación para que no se vea tan "escolar"
    x, y = rotate(x, y, np.deg2rad(12))
    x += 0.15
    y += 0.10

    ax.plot(x, y, color="black", lw=3.0, solid_capstyle="round")
    endpoint(ax, x[0], y[0])
    endpoint(ax, x[-1], y[-1])

    save_fig(fig, "geometry_lyapunov_first_order_cusps.png")


# ============================================================
# 3) Logarithmic star without cusps
# ============================================================

def draw_logarithmic_star_no_cusps():
    fig, ax = plt.subplots(figsize=(4.8, 3.0))
    setup_ax(ax, xlim=(-3.0, 3.0), ylim=(-2.6, 2.6))

    # tres ramas tipo espiral logarítmica
    base_angles = [np.deg2rad(25), np.deg2rad(150), np.deg2rad(280)]

    for phi0 in base_angles:
        s = np.linspace(0, 1, 500)
        r = 0.06 * np.exp(3.25 * s)
        theta = phi0 + 1.15 * s
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        ax.plot(x, y, color="black", lw=2.8, solid_capstyle="round")
        endpoint(ax, x[-1], y[-1], size=42)

    common_point(ax, 0.0, 0.0, size=32)
    save_fig(fig, "geometry_logarithmic_star_no_cusps.png")


# ============================================================
# 4) Pure cusp star-like curve
#    Varias ramas cuspidales encontrándose en el mismo punto.
# ============================================================

def cusp_branch(scale_x=2.2, scale_y=1.0, n=500):
    t = np.linspace(0, 1.75, n)
    x = scale_x * t**2
    y = scale_y * t**3
    return x, y


def draw_pure_cusp_star():
    fig, ax = plt.subplots(figsize=(4.8, 3.0))
    setup_ax(ax, xlim=(-3.0, 3.0), ylim=(-2.5, 2.5))

    angles = [np.deg2rad(20), np.deg2rad(105), np.deg2rad(205), np.deg2rad(300)]

    for ang in angles:
        x, y = cusp_branch(scale_x=0.85, scale_y=0.35)
        xr, yr = rotate(x, y, ang)
        ax.plot(xr, yr, color="black", lw=2.8, solid_capstyle="round")
        endpoint(ax, xr[-1], yr[-1], size=42)

    common_point(ax, 0.0, 0.0, size=34)
    save_fig(fig, "geometry_pure_cusp_star.png")


# ============================================================
# 5) Mixed cusp-angle star
#    Dos ramas cuspidales + dos rayos
# ============================================================

def draw_mixed_cusp_angle_star():
    fig, ax = plt.subplots(figsize=(4.8, 3.0))
    setup_ax(ax, xlim=(-3.2, 3.2), ylim=(-2.5, 2.5))

    # Dos ramas cuspidales a la derecha
    cusp_angles = [np.deg2rad(35), np.deg2rad(-35)]
    for ang in cusp_angles:
        x, y = cusp_branch(scale_x=0.90, scale_y=0.36)
        xr, yr = rotate(x, y, ang)
        ax.plot(xr, yr, color="black", lw=2.8, solid_capstyle="round")
        endpoint(ax, xr[-1], yr[-1], size=42)

    # Dos rayos a la izquierda
    ray_angles = [np.deg2rad(150), np.deg2rad(210)]
    for ang in ray_angles:
        r = np.linspace(0, 2.6, 200)
        x = r * np.cos(ang)
        y = r * np.sin(ang)
        ax.plot(x, y, color="black", lw=2.8, solid_capstyle="round")
        endpoint(ax, x[-1], y[-1], size=42)

    common_point(ax, 0.0, 0.0, size=34)
    save_fig(fig, "geometry_mixed_cusp_angle_star.png")


# ============================================================
# Ejecutar todo
# ============================================================

if __name__ == "__main__":
    draw_lyapunov_no_cusps()
    draw_lyapunov_first_order_cusps()
    draw_logarithmic_star_no_cusps()
    draw_pure_cusp_star()
    draw_mixed_cusp_angle_star()
```
