```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_points(x, y, theta):
    """
    Rotación rígida del plano por ángulo theta (en radianes),
    alrededor del origen:
        [u]   [cosθ  -sinθ][x]
        [v] = [sinθ   cosθ][y]
    """
    c, s = np.cos(theta), np.sin(theta)
    u = c * x - s * y
    v = s * x + c * y
    return u, v

def plot_before_after(theta=-np.pi/4, tmin=-2.0, tmax=2.0, n=2000):
    # Parametrización de y = cbrt(x): x=t^3, y=t
    t = np.linspace(tmin, tmax, n)
    x = t**3
    y = t

    # Rotación
    u, v = rotate_points(x, y, theta)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # --- Panel 1: Original ---
    ax = axes[0]
    ax.plot(x, y, linewidth=2)
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.set_title(r"Antes: $y=\sqrt[3]{x}$ (tangente vertical en $0$)")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_aspect("equal", adjustable="box")

    # Línea vertical de referencia cerca del origen
    ax.plot([0, 0], [tmin, tmax], linestyle="--", linewidth=1)

    # --- Panel 2: Rotada ---
    ax = axes[1]
    ax.plot(u, v, linewidth=2)
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.set_title(r"Después: curva rotada por $\theta={:.3f}$ rad".format(theta))
    ax.set_xlabel("u")
    ax.set_ylabel("v")
    ax.set_aspect("equal", adjustable="box")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Cambia theta si quieres otra rotación
    plot_before_after(theta=-np.pi/4, tmin=-2, tmax=2, n=3000)
```
